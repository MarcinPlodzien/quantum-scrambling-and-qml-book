# Exercise 7.7 — Barren Plateaus through the Lens of Quantum Fisher Information

**Chapter 7: Quantum Machine Learning** &nbsp;|&nbsp; *Section 7.2: Barren Plateaus and Trainability*

---

This exercise reveals the deep geometric origin of barren plateaus by connecting the trainability of parameterized quantum circuits to the **Quantum Fisher Information (QFI)** with respect to variational parameters. We show that the QFI itself concentrates exponentially for Haar-random circuits, providing a parameter-space geometric explanation for why gradients vanish.

## 1. The Geometric Picture

Consider a parameterized circuit $U(\boldsymbol{\theta})$ preparing the state $|\psi(\boldsymbol{\theta})\rangle = U(\boldsymbol{\theta})|0\rangle$. The cost function is $C(\boldsymbol{\theta}) = \langle \psi(\boldsymbol{\theta}) | O | \psi(\boldsymbol{\theta}) \rangle$ for some observable $O$.

The **Quantum Fisher Information** (QFI) with respect to the parameter $\theta_k$ measures how sensitively the quantum state responds to infinitesimal changes in $\theta_k$:

$$
F_Q(\theta_k) = 4 \left( \langle \partial_k \psi | \partial_k \psi \rangle - |\langle \psi | \partial_k \psi \rangle|^2 \right).
$$

This is precisely **four times the Fubini-Study metric** $g_{kk}^{\mathrm{FS}}$ on the projective Hilbert space $\mathbb{CP}^{D-1}$:

$$
F_Q(\theta_k) = 4\, g_{kk}^{\mathrm{FS}}.
$$

When $F_Q(\theta_k) \to 0$, the state $|\psi\rangle$ becomes **insensitive** to $\theta_k$: an infinitesimal change in $\theta_k$ produces a negligible change in the quantum state. If the state doesn't change, the cost function doesn't change, and the gradient vanishes.

This provides a purely geometric, observable-independent explanation of barren plateaus: the parameter landscape becomes **metrically flat** in the Fubini-Study sense.

## 2. Mathematical Toolbox

**Tool 1 — QFI for pure states:** For $|\psi(\theta)\rangle = e^{-i\theta G} |\psi_0\rangle$ with generator $G$,
$$
F_Q(\theta) = 4\, \mathrm{Var}_{|\psi_0\rangle}(G) = 4\left(\langle G^2 \rangle - \langle G \rangle^2\right).
$$

**Tool 2 — Cramér-Rao bound on gradients:** The variance of $\partial C / \partial \theta_k$ over the parameter distribution is bounded by:
$$
\mathrm{Var}\!\left[\frac{\partial C}{\partial \theta_k}\right] \leq \|O\|^2 \cdot F_Q(\theta_k).
$$
If $F_Q \to 0$, the gradient is forced to vanish regardless of the observable $O$.

**Tool 3 — Haar first moment (Weingarten $k=1$):**
$$
\mathbb{E}_{U}[U^\dagger A\, U] = \frac{\mathrm{Tr}(A)}{D}\, I.
$$

**Tool 4 — Haar second moment (Weingarten $k=2$):**
$$
\mathbb{E}_{U}[U^\dagger A\, U \otimes U^\dagger B\, U] = \alpha(A,B)\, I \otimes I + \beta(A,B)\, \mathrm{SWAP},
$$
where $\alpha = \frac{D\,\mathrm{Tr}(A)\mathrm{Tr}(B) - \mathrm{Tr}(AB)}{D(D^2-1)}$ and $\beta = \frac{D\,\mathrm{Tr}(AB) - \mathrm{Tr}(A)\mathrm{Tr}(B)}{D(D^2-1)}$.

## 3. Exercise Statement

Consider a circuit $U(\theta) = V_2\, e^{-i\theta G}\, V_1$ where $V_1, V_2$ are Haar-random and $G$ is a fixed Hermitian generator (e.g. a single Pauli).

**(a)** Show that the QFI with respect to $\theta$ is $F_Q(\theta) = 4\,\mathrm{Var}_{|\psi_1\rangle}(G)$ where $|\psi_1\rangle = V_1|0\rangle$.

**(b)** Show that for Haar-random $V_1$, the expected QFI is:
$$
\mathbb{E}_{V_1}[F_Q] = \frac{4}{D+1}\left(\mathrm{Tr}(G^2) - \frac{(\mathrm{Tr}\, G)^2}{D}\right).
$$

**(c)** For a traceless Pauli generator $G$ with $G^2 = I$ and $\mathrm{Tr}(G) = 0$, show that $\mathbb{E}[F_Q] = \frac{4D}{D+1} \xrightarrow{D \to \infty} 4$. Compare this with the gradient variance $\sim N/D$.

**(d)** Now consider the **effective QFI** $\tilde{F}_Q = F_Q(\theta) \cdot \|\partial_\theta \langle O \rangle / F_Q\|^2$. Show that the gradient variance is bounded by $\mathrm{Var}[\partial C/\partial\theta] \leq \|O\|^2 \cdot \mathbb{E}[F_Q] / D^2$, giving the exponential suppression.

## 4. Step-by-Step Derivation

### Part (a): QFI from the variance of the generator

At $\theta = 0$, the state is $|\psi_1\rangle = V_1|0\rangle$ and the parameter acts via $e^{-i\theta G}$. The QFI formula for pure states gives:

$$
F_Q = 4\left(\langle \psi_1 | G^2 | \psi_1 \rangle - \langle \psi_1 | G | \psi_1 \rangle^2\right) = 4\,\mathrm{Var}_{|\psi_1\rangle}(G). \quad \checkmark
$$

Note: the post-rotation $V_2$ does not affect $F_Q$ because the Fubini-Study metric is invariant under global unitaries. The QFI captures the intrinsic sensitivity of the state manifold, not the measurement basis.

### Part (b): Haar-averaged QFI

We need $\mathbb{E}_{V_1}[\langle\psi_1|G^2|\psi_1\rangle]$ and $\mathbb{E}_{V_1}[\langle\psi_1|G|\psi_1\rangle^2]$.

**First term** (using Tool 3, $k=1$):
$$
\mathbb{E}[\langle G^2 \rangle] = \mathrm{Tr}\!\left(\frac{I}{D} G^2\right) = \frac{\mathrm{Tr}(G^2)}{D}.
$$

**Second term** (using Tool 4, $k=2$): Since $\langle G \rangle^2 = \langle \psi_1 | G | \psi_1 \rangle^2$, this is a second-moment integral:
$$
\mathbb{E}[\langle G \rangle^2] = \frac{\mathrm{Tr}(G)^2 + \mathrm{Tr}(G^2)}{D(D+1)}.
$$

Combining:
$$
\mathbb{E}[F_Q] = 4\left(\frac{\mathrm{Tr}(G^2)}{D} - \frac{\mathrm{Tr}(G)^2 + \mathrm{Tr}(G^2)}{D(D+1)}\right) = \frac{4}{D+1}\left(\mathrm{Tr}(G^2) - \frac{\mathrm{Tr}(G)^2}{D}\right). \quad \checkmark
$$

### Part (c): Pauli generator and the trainability paradox

For a traceless Pauli $G$ ($G^2 = I$, $\mathrm{Tr}(G) = 0$, $\mathrm{Tr}(G^2) = D$):

$$
\mathbb{E}[F_Q] = \frac{4D}{D+1} \xrightarrow{D \to \infty} 4.
$$

This is an $\mathcal{O}(1)$ constant! The QFI does **not** vanish exponentially. The quantum state *is* sensitive to the parameter $\theta$.

But recall that the gradient variance $\mathrm{Var}[\partial C / \partial \theta] \sim N/D \to 0$ exponentially. This seems contradictory: the state moves in Hilbert space, but the cost doesn't change?

**Resolution:** The state $|\psi(\theta)\rangle$ moves along a geodesic in $\mathbb{CP}^{D-1}$, but the observable $O$ projects onto a tiny subspace. The motion is "perpendicular" to the cost landscape. The QFI tells us the state moves; the barren plateau tells us it moves in a direction that doesn't matter for the observable. This is the **curse of expressibility**: the circuit explores exponentially many directions, but only $\mathcal{O}(1)$ of them align with a given cost function.

### Part (d): The observable-weighted QFI bound

The gradient is $\partial_\theta C = 2\,\mathrm{Im}\langle \psi | O | \partial_\theta \psi \rangle$. By the extended Cramér-Rao inequality:

$$
\mathrm{Var}[\partial_\theta C] \leq \mathrm{Var}[O] \cdot F_Q.
$$

For Haar-random states, $\mathrm{Var}[O] \sim \mathrm{Tr}(O^2)/D \sim 1/D$ (the observable variance itself concentrates). Combined with $F_Q \sim \mathcal{O}(1)$:

$$
\mathrm{Var}[\partial_\theta C] \lesssim \frac{\mathrm{Tr}(O^2)}{D} \cdot \frac{4D}{D+1} \sim \frac{4\,\mathrm{Tr}(O^2)}{D}.
$$

For a global observable with $\mathrm{Tr}(O^2) = \mathcal{O}(1)$, this gives exponential suppression $\sim 1/D$.

## 5. Physical Interpretation: Geometry vs Trainability

The QFI analysis reveals a beautiful decomposition of the barren plateau phenomenon:

| Quantity | Scaling | Meaning |
|:--|:--|:--|
| $F_Q(\theta)$ | $\mathcal{O}(1)$ | The state *does* move in Hilbert space |
| $\mathrm{Var}_{|\psi\rangle}(O)$ | $\mathcal{O}(1/D)$ | The observable *cannot distinguish* nearby states |
| $\mathrm{Var}[\partial C/\partial\theta]$ | $\mathcal{O}(1/D)$ | The cost landscape is **flat** |

The barren plateau is not a failure of the parameterization to move the state—it is a failure of the observable to *detect* the motion. The Fubini-Study geodesic ball explored by the parameter $\theta$ has an exponentially large volume, but the "shadow" of this ball onto the cost function is exponentially thin.

This motivates two strategies:
1. **Local observables:** If $O$ acts on $k$ qubits, $\mathrm{Tr}(O^2) \sim 2^k$ and $\mathrm{Var}[\partial C] \sim 2^k / D$, which is polynomial for $k = \mathcal{O}(\log N)$.
2. **Structured ans\"atze:** Restrict the circuit to a low-dimensional submanifold of $\mathbb{CP}^{D-1}$ where $F_Q$ and the cost function are aligned.

## 6. Symbolic Verification (SymPy)

```python
import sympy as sp

D = sp.Symbol('D', positive=True, integer=True)

# Part (b): E[F_Q] for general generator
TrG2 = sp.Symbol('TrG2', positive=True)  # Tr(G^2)
TrG = sp.Symbol('TrG')                    # Tr(G)

E_G2 = TrG2 / D
E_G_sq = (TrG**2 + TrG2) / (D * (D + 1))

E_FQ = 4 * sp.simplify(E_G2 - E_G_sq)
E_FQ_simplified = sp.simplify(E_FQ)
print(f'E[F_Q] = {E_FQ_simplified}')

# Verify matches expected form: 4/(D+1) * (Tr(G^2) - Tr(G)^2/D)
expected = sp.Rational(4,1) / (D + 1) * (TrG2 - TrG**2 / D)
assert sp.simplify(E_FQ_simplified - expected) == 0
print(f'Verified: E[F_Q] = 4/(D+1) * (Tr(G^2) - Tr(G)^2/D)')

# Part (c): Pauli generator: Tr(G) = 0, Tr(G^2) = D
E_FQ_pauli = E_FQ_simplified.subs([(TrG, 0), (TrG2, D)])
print(f'\nPauli generator: E[F_Q] = {sp.simplify(E_FQ_pauli)}')
print(f'Large D limit: {sp.limit(E_FQ_pauli, D, sp.oo)}')

# Part (d): Gradient bound
TrO2 = sp.Symbol('TrO2', positive=True)
Var_grad_bound = TrO2 / D * E_FQ_pauli
print(f'\nGradient variance bound: {sp.simplify(Var_grad_bound)}')
print(f'Large D: ~ {sp.limit(Var_grad_bound, D, sp.oo)} * Tr(O^2) / D')
```

## 7. Numerical Verification (NumPy)

```python
import numpy as np
from scipy.stats import unitary_group

np.random.seed(42)

print('QFI Concentration for Haar-Random States')
print('=' * 60)
print(f'{"N":>3s}  {"D":>5s}  {"E[F_Q] MC":>10s}  {"4D/(D+1)":>10s}  {"Var[dC] MC":>11s}  {"N/D":>8s}')
print('-' * 60)

for N in [2, 3, 4, 5, 6]:
    D = 2**N
    # Generator: Z on qubit 1
    G = np.diag([(-1)**int(b) for b in range(D)])
    # Observable: Z on qubit 1 (same for simplicity)
    O = G.copy()
    
    fq_list = []
    grad_list = []
    
    for _ in range(3000):
        V1 = unitary_group.rvs(D)
        psi = V1[:, 0]
        
        # QFI = 4 * Var(G)
        eG = (psi.conj() @ G @ psi).real
        eG2 = (psi.conj() @ G @ G @ psi).real
        fq = 4 * (eG2 - eG**2)
        fq_list.append(fq)
        
        # Gradient: dC/dtheta = 2*Im(<psi|O * G|psi>)
        grad = 2 * np.imag(psi.conj() @ O @ G @ psi)
        grad_list.append(grad)
    
    mc_fq = np.mean(fq_list)
    th_fq = 4 * D / (D + 1)
    mc_var = np.var(grad_list)
    th_var_approx = N / D
    
    print(f'{N:3d}  {D:5d}  {mc_fq:10.4f}  {th_fq:10.4f}  {mc_var:11.6f}  {th_var_approx:8.6f}')

print('\nKey insight: F_Q ~ O(1) but Var[grad] ~ O(1/D)!')
print('The state moves, but the cost cannot detect the motion.')
```

## 8. Visual Proof: QFI vs Gradient Scaling

```python
import matplotlib.pyplot as plt

N_vals = np.arange(2, 12)
D_vals = 2**N_vals

FQ_analytical = 4 * D_vals / (D_vals + 1)
grad_var = N_vals / D_vals

plt.rcParams['figure.dpi'] = 150
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4.5))

# Left: QFI stays O(1)
ax1.plot(N_vals, FQ_analytical, 'ko-', markersize=7, linewidth=2, label=r'$\mathbb{E}[F_Q]$')
ax1.axhline(4, color='crimson', linestyle='--', linewidth=1.5, label=r'Limit $\to 4$')
ax1.set_xlabel('Number of qubits $N$', fontsize=11)
ax1.set_ylabel(r'$\mathbb{E}[F_Q(\theta)]$', fontsize=11)
ax1.set_title('QFI: State Sensitivity (stays $\mathcal{O}(1)$)', fontsize=12)
ax1.legend(fontsize=9)
ax1.grid(True, ls='--', alpha=0.4)
ax1.set_ylim(0, 5)

# Right: gradient variance dies exponentially
ax2.semilogy(N_vals, grad_var, 'ko-', markersize=7, linewidth=2, label=r'$\mathrm{Var}[\partial C / \partial\theta]$')
ax2.semilogy(N_vals, 1.0 / D_vals, 'r--', linewidth=1.5, label=r'$1/D = 2^{-N}$')
ax2.set_xlabel('Number of qubits $N$', fontsize=11)
ax2.set_ylabel('Gradient variance', fontsize=11)
ax2.set_title('Barren Plateau: Cost Insensitivity', fontsize=12)
ax2.legend(fontsize=9)
ax2.grid(True, which='both', ls='--', alpha=0.4)

plt.suptitle('The Trainability Paradox: Motion without Detection', fontsize=13, y=1.03)
plt.tight_layout()
plt.show()
```
