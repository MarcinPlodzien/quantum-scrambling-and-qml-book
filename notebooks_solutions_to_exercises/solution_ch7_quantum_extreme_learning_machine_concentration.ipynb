This exercise develops the complete theory of why deep quantum circuits destroy the utility of Quantum Extreme Learning Machines. We begin with the physical picture, formalize the POVM structure, and then prove three escalating results about signal death under Haar scrambling.

## 1. The Physical Setup

Consider a quantum device with two registers:

- An **input register** of $N_{\mathrm{in}}$ qubits, encoding classical data $x$ via a state $\rho_{\mathrm{in}}(x)$.
- A **reservoir register** of $N_{\mathrm{res}}$ qubits, initialized in the fiducial state $|0\rangle^{\otimes N_{\mathrm{res}}}$.

The total Hilbert space dimension is $D = 2^{N_{\mathrm{in}} + N_{\mathrm{res}}}$. A fixed (but random) unitary $U$ acts on the joint system, and features are extracted by measuring in the computational basis:

$$
f_b(x) = \mathrm{Tr}\!\big(|b\rangle\langle b|\, U\, \sigma(x)\, U^\dagger\big), \qquad \sigma(x) = \rho_{\mathrm{in}}(x) \otimes |0\rangle\langle 0|_{\mathrm{res}}.
$$

The $D$ real numbers $\{f_b(x)\}_{b=0}^{D-1}$ form the feature vector that a classical linear model is trained on. The central question is: *how much information about $x$ survives in these features when $U$ is a deep, chaotic circuit?*

## 2. The POVM Perspective: Naimark Dilation in Reverse

A powerful way to understand the QELM is to recognize that the entire pipeline—tensoring with the reservoir, applying $U$, and projecting onto $|b\rangle$—defines an **effective measurement** on the input register alone.

Tracing over the reservoir degrees of freedom, each computational basis outcome $b$ corresponds to a **POVM element** acting on the input space $\mathcal{H}_{\mathrm{in}}$:

$$
M_b \;=\; \langle 0|_{\mathrm{res}}\, U^\dagger\, |b\rangle\langle b|\, U\, |0\rangle_{\mathrm{res}} \;\in\; \mathcal{L}(\mathcal{H}_{\mathrm{in}}).
$$

These satisfy the POVM axioms:
1. **Positivity:** $M_b \geq 0$ (since $M_b = A_b^\dagger A_b$ where $A_b = \langle b| U |0\rangle_{\mathrm{res}}$).
2. **Completeness:** $\sum_{b=0}^{D-1} M_b = I_{\mathrm{in}}$ (since $\{|b\rangle\}$ is a complete basis and $U$ is unitary).
3. **Informativeness:** Each $M_b$ has rank at most $D_{\mathrm{res}} = 2^{N_{\mathrm{res}}}$, so the POVM is **informationally complete** when $D_{\mathrm{res}} \geq D_{\mathrm{in}}$.

The feature is simply the Born probability under this POVM: $f_b(x) = \mathrm{Tr}(M_b\, \rho_{\mathrm{in}}(x))$.

This is precisely the **Naimark dilation theorem** run in reverse: a projective measurement on a larger space realizes a generalized POVM on the subsystem.

## 3. Mathematical Toolbox

The following results from Haar integration and Weingarten calculus are needed:

**Tool 1 — First Moment (Weingarten $k=1$):** For any operator $A$ on $\mathcal{H}$ with $\dim \mathcal{H} = D$,
$$
\mathbb{E}_{U \sim \mathrm{Haar}}\!\big[U^\dagger A\, U\big] = \frac{\mathrm{Tr}(A)}{D}\, I_D.
$$
This follows from Schur's lemma: the integral must commute with all unitaries, hence must be proportional to the identity. The coefficient is fixed by taking the trace of both sides.

**Tool 2 — Second Moment (Weingarten $k=2$):** The collision probability uses the $k=2$ Weingarten formula. For a pure state $|\psi\rangle$,
$$
\mathbb{E}_U\!\left[\sum_{b=0}^{D-1} |\langle b| U |\psi\rangle|^4 \right] = \frac{\mathrm{Tr}(\rho^2) + (\mathrm{Tr}\rho)^2}{D(D+1)} = \frac{2}{D+1},
$$
where we used $\mathrm{Tr}(\rho^2) = 1$ and $\mathrm{Tr}(\rho) = 1$ for a pure state. The Weingarten functions for $S_2$ are $\mathrm{Wg}(\mathrm{id}, D) = 1/(D^2 - 1)$ and $\mathrm{Wg}((12), D) = -1/(D(D^2 - 1))$.

**Tool 3 — Haar Symmetry:** The Haar measure is invariant under basis permutations: $\mathbb{E}[f_b^2]$ is independent of $b$, so $\mathbb{E}[f_b^2] = \frac{1}{D} \mathbb{E}[\sum_b f_b^2]$.

**Tool 4 — Shot Noise Statistics:** Estimating a probability $p$ from $\nu$ Bernoulli trials gives a standard error $\sigma_{\mathrm{shot}} = \sqrt{p(1-p)/\nu}$.

## 4. Exercise Statement

Assume $U$ is drawn from the Haar measure on $U(D)$ and the input state is pure.

**(a) Signal Death:** Show that the Haar-averaged POVM element is $\mathbb{E}_U[M_b] = \frac{1}{D} I_{\mathrm{in}}$, and consequently that $\mathbb{E}_U[f_b(x)] = 1/D$ for all inputs $x$.

**(b) Speckle Variance:** Show that the feature variance is $\mathrm{Var}_U[f_b] = \frac{D-1}{D^2(D+1)}$.

**(c) Exponential Shot Cost:** Show that resolving the input-dependent signal from measurement shot noise requires $\nu > D = 2^N$ shots.

## 5. Step-by-Step Derivation

### Part (a): The Haar average trivializes the POVM

We compute the average of each POVM element over the Haar measure. Using Tool 1 (Weingarten $k=1$):

$$
\mathbb{E}_U[M_b] = \langle 0|_{\mathrm{res}}\, \mathbb{E}_U\!\big[U^\dagger |b\rangle\langle b| U\big]\, |0\rangle_{\mathrm{res}} = \langle 0|_{\mathrm{res}}\, \frac{I_D}{D}\, |0\rangle_{\mathrm{res}} = \frac{1}{D}\, I_{\mathrm{in}}.
$$

On average, every POVM element $M_b$ is proportional to the identity on the input space. It carries **zero directional information** about the input state. Every input produces the same Born probability:

$$
\mathbb{E}_U[f_b(x)] = \mathrm{Tr}\!\left(\frac{I_{\mathrm{in}}}{D}\, \rho_{\mathrm{in}}(x)\right) = \frac{1}{D}.
$$

Geometrically, in the space of POVM elements, the Haar average projects every $M_b$ onto the unique $\mathrm{SU}(D)$-invariant point: the normalized identity. This is the POVM-theoretic manifestation of *information scrambling*.

### Part (b): Speckle fluctuations carry the signal

The information about the input $x$ lives entirely in the fluctuations of $M_b$ around its mean $I_{\mathrm{in}}/D$. Using Tool 2 (Weingarten $k=2$), the collision probability for a pure input is:

$$
\mathbb{E}_U\!\left[\sum_b f_b^2\right] = \frac{2}{D+1}.
$$

By Tool 3 (Haar symmetry), $\mathbb{E}_U[f_b^2] = \frac{2}{D(D+1)}$.

The variance of each feature is:

$$
\mathrm{Var}_U[f_b] = \frac{2}{D(D+1)} - \frac{1}{D^2} = \frac{2D - (D+1)}{D^2(D+1)} = \frac{D-1}{D^2(D+1)}.
$$

For large $D$, $\mathrm{Var}[f_b] \sim 1/D^2$. The useful features are exponentially small perturbations on top of the uniform background $1/D$.

### Part (c): Exponential measurement cost

By Tool 4, the shot noise from $\nu$ measurements is $\sigma_{\mathrm{shot}} \approx 1/\sqrt{\nu D}$. For the signal ($\sim 1/D$) to dominate:

$$
\frac{1}{D} > \frac{1}{\sqrt{\nu D}} \implies \nu > D = 2^{N_{\mathrm{in}} + N_{\mathrm{res}}}.
$$

An exponential shot budget is required simply to **detect** the presence of an input-dependent signal.

## 6. Physical Interpretation: The Information-Extraction Hierarchy

| Scrambling Regime | POVM Quality | Feature Variance | Shot Cost | ML Utility |
|:--|:--|:--|:--|:--|
| None (identity $U$) | Projective on input | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ | Input-limited |
| Partial (finite depth) | Structured, rank-deficient | $\mathcal{O}(1/\mathrm{poly}(D))$ | $\mathrm{poly}(D)$ | Useful |
| Full (Haar random) | Trivial ($\propto I$) | $\mathcal{O}(1/D^2)$ | $\mathcal{O}(D)$ | Useless |

The sweet spot for quantum machine learning lies at the **edge of scrambling**: deep enough that the POVM elements explore a rich subspace, but shallow enough that they retain directional sensitivity to the input.

## 7. Symbolic Verification (SymPy)

```python
import sympy as sp

D = sp.Symbol('D', positive=True, integer=True)

# --- Part (a): Expected feature value ---
E_f = sp.Rational(1, 1) / D
print(f'Part (a): E[f_b] = {E_f}')

# --- Part (b): Variance derivation ---
# Second moment from collision probability
E_f2 = sp.Rational(2, 1) / (D * (D + 1))
Var_f = sp.simplify(E_f2 - E_f**2)
expected_var = (D - 1) / (D**2 * (D + 1))

print(f'Part (b): E[f^2] = {E_f2}')
print(f'          Var[f] = {Var_f}')
assert sp.simplify(Var_f - expected_var) == 0, 'Variance mismatch!'
print(f'          Verified: (D-1)/[D^2(D+1)]')

# Asymptotics
print(f'\nAsymptotic expansion for large D:')
print(f'  E[f]   = {sp.series(E_f, D, sp.oo, 2)}')
print(f'  Var[f] = {sp.series(Var_f, D, sp.oo, 4)}')

# --- Part (c): Shot threshold ---
nu = sp.Symbol('nu', positive=True)
shot_noise_sq = E_f / nu  # sigma_shot^2 ~ f/nu ~ 1/(nu*D)
signal_sq = Var_f          # sigma_f^2
# Require signal > shot noise: Var_f > 1/(nu*D)
threshold = sp.solve(signal_sq - shot_noise_sq, nu)
print(f'\nPart (c): Shot threshold nu > {sp.simplify(threshold[0])}')
print(f'  For large D: nu > {sp.limit(threshold[0], D, sp.oo)} * D')
```

## 8. Numerical Verification (NumPy)

```python
import numpy as np
from scipy.stats import unitary_group

np.random.seed(42)

print('QELM POVM Feature Concentration')
print('=' * 70)
print(f'{"Config":>8s}  {"D":>4s}  {"E[f] MC":>9s}  {"1/D":>9s}  {"Var MC":>10s}  {"Var Exact":>10s}')
print('-' * 70)

for Nin, Nres in [(1,1), (1,2), (2,2)]:
    N = Nin + Nres; D = 2**N
    D_in = 2**Nin; D_res = 2**Nres

    psi_in = np.ones(D_in) / np.sqrt(D_in)
    psi_res = np.zeros(D_res); psi_res[0] = 1
    sigma = np.kron(np.outer(psi_in, psi_in), np.outer(psi_res, psi_res))

    features = []
    for _ in range(5000):
        U = unitary_group.rvs(D)
        rho = U @ sigma @ U.conj().T
        features.append(np.diag(rho).real)

    features = np.array(features)
    mean_f = np.mean(features)
    var_f = np.mean(np.var(features, axis=0))
    var_exact = (D - 1) / (D**2 * (D + 1))

    print(f'{Nin}+{Nres}: {D:4d}  {mean_f:9.5f}  {1/D:9.5f}  {var_f:10.2e}  {var_exact:10.2e}')
    assert abs(mean_f - 1/D) < 0.01
    assert abs(var_f - var_exact) / var_exact < 0.15

print('\nAll assertions passed.')
```

## 9. Visual Proof: Porter-Thomas Speckle Distribution

```python
import matplotlib.pyplot as plt

D_plot = 16
fb_samples = []
for _ in range(500):
    U = unitary_group.rvs(D_plot)
    psi = U[:, 0]
    fb_samples.append(abs(psi[0])**2)

plt.rcParams['figure.dpi'] = 150
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4.5))

ax1.hist(fb_samples, bins=40, color='royalblue', alpha=0.7, density=True,
         edgecolor='white', linewidth=0.5)
p_grid = np.linspace(0, max(fb_samples)*1.1, 200)
ax1.plot(p_grid, D_plot * np.exp(-D_plot * p_grid), 'k-', linewidth=2,
         label=r'Porter-Thomas $De^{-Df}$')
ax1.axvline(1/D_plot, color='crimson', linestyle='--', linewidth=2,
            label=r'$1/D$')
ax1.set_xlabel(r'Feature value $f_b$', fontsize=11)
ax1.set_ylabel('Probability density', fontsize=11)
ax1.set_title(f'Single Feature Distribution (D={D_plot})', fontsize=12)
ax1.legend(fontsize=9)
ax1.grid(True, ls='--', alpha=0.4)

D_range = 2**np.arange(2, 12)
var_exact = (D_range - 1) / (D_range**2 * (D_range + 1))
var_approx = 1.0 / D_range**2
ax2.loglog(D_range, var_exact, 'ko-', markersize=6, linewidth=2,
           label=r'Exact $(D-1)/[D^2(D+1)]$')
ax2.loglog(D_range, var_approx, 'r--', linewidth=1.5,
           label=r'Asymptotic $1/D^2$')
ax2.set_xlabel(r'Hilbert space dimension $D$', fontsize=11)
ax2.set_ylabel(r'Feature variance', fontsize=11)
ax2.set_title('Signal Death under Scrambling', fontsize=12)
ax2.legend(fontsize=9)
ax2.grid(True, which='both', ls='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
