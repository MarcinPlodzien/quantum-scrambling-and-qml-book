## Motivation

Pure quantum states live not in Hilbert space $\mathcal{H}$ but in **projective** Hilbert space $\mathbb{CP}^{D-1}$ (since global phases are unphysical).  The natural metric on this space is the **Fubini-Study metric**, which measures the intrinsic distance between quantum states â€” the same quantity that controls the quantum speed limit and the quantum geometric tensor.  For a single qubit, $\mathbb{CP}^1 \cong S^2$: the Fubini-Study metric gives the Bloch sphere its geometry.

## Exercise Statement

A general single-qubit pure state is

$$
|\psi(\theta, \varphi)\rangle = \cos\frac{\theta}{2}|0\rangle + e^{i\varphi}\sin\frac{\theta}{2}|1\rangle.
$$

**(a)** Starting from the Fubini-Study line element $ds^2_{\mathrm{FS}} = \langle d\psi|d\psi\rangle - |\langle\psi|d\psi\rangle|^2$, show that

$$
ds^2_{\mathrm{FS}} = \frac{1}{4}\bigl(d\theta^2 + \sin^2\!\theta\, d\varphi^2\bigr).
$$

**(b)** Compute the geodesic distance between $|0\rangle$ (north pole) and $|+\rangle = (|0\rangle + |1\rangle)/\sqrt{2}$ (equator).

## Solution

### Part (a): Deriving the metric

**Step 1: Compute $|d\psi\rangle$.**

Differentiating the state:

$$
|d\psi\rangle = -\frac{1}{2}\sin\frac{\theta}{2}\,d\theta\,|0\rangle + e^{i\varphi}\left[\frac{1}{2}\cos\frac{\theta}{2}\,d\theta + i\sin\frac{\theta}{2}\,d\varphi\right]|1\rangle.
$$

**Step 2: Compute $\langle d\psi|d\psi\rangle$ (Term 1).**

The squared norm gives three contributions:

$$
\langle d\psi|d\psi\rangle = \frac{1}{4}\sin^2\frac{\theta}{2}\,d\theta^2 + \frac{1}{4}\cos^2\frac{\theta}{2}\,d\theta^2 + \sin^2\frac{\theta}{2}\,d\varphi^2.
$$

Using $\sin^2 + \cos^2 = 1$:

$$
\langle d\psi|d\psi\rangle = \frac{1}{4}d\theta^2 + \sin^2\frac{\theta}{2}\,d\varphi^2.
$$

**Step 3: Compute $|\langle\psi|d\psi\rangle|^2$ (Term 2).**

The inner product $\langle\psi|d\psi\rangle$ has two contributions:

- From $d\theta$: $\cos\frac{\theta}{2}\cdot(-\frac{1}{2}\sin\frac{\theta}{2}) + \sin\frac{\theta}{2}\cdot\frac{1}{2}\cos\frac{\theta}{2} = 0$. (The real part cancels!)
- From $d\varphi$: $\sin\frac{\theta}{2}\cdot i\sin\frac{\theta}{2} = i\sin^2\frac{\theta}{2}\,d\varphi$.

Therefore:

$$
|\langle\psi|d\psi\rangle|^2 = \sin^4\frac{\theta}{2}\,d\varphi^2.
$$

**Step 4: Combine.**

$$
ds^2_{\mathrm{FS}} = \frac{1}{4}d\theta^2 + \sin^2\frac{\theta}{2}\,d\varphi^2 - \sin^4\frac{\theta}{2}\,d\varphi^2 = \frac{1}{4}d\theta^2 + \sin^2\frac{\theta}{2}\cos^2\frac{\theta}{2}\,d\varphi^2.
$$

Using $\sin^2\frac{\theta}{2}\cos^2\frac{\theta}{2} = \frac{1}{4}\sin^2\theta$:

$$
\boxed{ds^2_{\mathrm{FS}} = \frac{1}{4}\bigl(d\theta^2 + \sin^2\!\theta\,d\varphi^2\bigr).}
$$

This is the metric of a **round sphere of radius** $R = 1/2$.  The Bloch sphere has half the radius of the "geographic" unit sphere â€” the factor $1/2$ arises from the half-angle parametrization of $\mathrm{SU}(2)$.

### Part (b): Geodesic distance $|0\rangle \to |+\rangle$

The north pole is $\theta = 0$ and $|+\rangle$ is at $\theta = \pi/2$, $\varphi = 0$.  Along the meridian ($d\varphi = 0$), the geodesic length is:

$$
d_{\mathrm{FS}} = \int_0^{\pi/2} \frac{1}{2}\,d\theta = \frac{\pi}{4}.
$$

Equivalently, using the closed-form Fubini-Study distance:

$$
d_{\mathrm{FS}}(|\psi\rangle, |\phi\rangle) = \arccos|\langle\psi|\phi\rangle|.
$$

$$
|\langle 0|+\rangle| = \frac{1}{\sqrt{2}}, \qquad d_{\mathrm{FS}} = \arccos\frac{1}{\sqrt{2}} = \frac{\pi}{4}. \quad \checkmark
$$

## Symbolic Verification (SymPy)

```python
import sympy as sp

t, p = sp.symbols('theta phi', real=True)

# Term 1: <dÏˆ|dÏˆ>
term1_dtheta = sp.Rational(1, 4)
term1_dphi = sp.sin(t/2)**2

# Term 2: |<Ïˆ|dÏˆ>|Â²
term2_dphi = sp.sin(t/2)**4

# FS metric coefficient for dÏ†Â²
g_phi = sp.trigsimp(term1_dphi - term2_dphi)
print(f"g_Ï†Ï† = {g_phi}")
print(f"g_Ï†Ï† = (1/4)sinÂ²Î¸ : {sp.trigsimp(g_phi - sp.sin(t)**2/4) == 0}")

# Geodesic distance
dist = sp.acos(1/sp.sqrt(2))
print(f"d_FS(|0âŸ©, |+âŸ©) = {dist} = {sp.nsimplify(dist)}")
```

## Numerical Verification

```python
import numpy as np

# Verify d_FS via the overlap formula for several pairs
ket0 = np.array([1, 0], dtype=complex)
ket1 = np.array([0, 1], dtype=complex)
ket_plus = np.array([1, 1], dtype=complex) / np.sqrt(2)

pairs = [
    ("|0âŸ©", "|+âŸ©", ket0, ket_plus, np.pi/4),
    ("|0âŸ©", "|1âŸ©", ket0, ket1, np.pi/2),
    ("|0âŸ©", "|0âŸ©", ket0, ket0, 0.0),
]

for name1, name2, psi, phi, expected in pairs:
    overlap = abs(psi.conj() @ phi)
    d_fs = np.arccos(np.clip(overlap, 0, 1))
    print(f"d_FS({name1}, {name2}) = {d_fs:.4f} (expected {expected:.4f})")
    assert np.isclose(d_fs, expected, atol=1e-10)

print("\nAll assertions passed. âœ“")
```
