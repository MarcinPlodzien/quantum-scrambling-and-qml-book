# Exercise 1.1 — Trace Distance and Fidelity for Two Qubit States

## Motivation

Two states can be compared in more than one way, and which comparison is the right one depends on the physical question.  The **trace distance** $T(\rho,\sigma) = \frac{1}{2}\|\rho-\sigma\|_1$ fixes the best single-shot probability of telling the states apart, $p_{\mathrm{correct}} = \frac{1}{2}(1+T)$, which is the content of the Helstrom bound.  The **Uhlmann fidelity** $F(\rho,\sigma)$ measures closeness through purifications instead, and it is the quantity that surfaces in randomized benchmarking and in the Bures geometry of state space.

The two are tied together by the inequalities $1-\sqrt{F} \leq T \leq \sqrt{1-F}$, which say that the two notions of closeness can never disagree qualitatively.  A pair of pure qubit states is the cleanest setting in which both quantities are available in closed form, so here the inequalities can be examined term by term instead of merely invoked.

## Exercise Statement

Take the two qubit states $\rho = |0\rangle\langle 0|$ and $\sigma = |\psi\rangle\langle\psi|$ with

$$
|\psi\rangle = \cos(\alpha/2)\,|0\rangle + \sin(\alpha/2)\,|1\rangle .
$$

**(a)** Compute the trace distance $T(\rho,\sigma) = \frac{1}{2}\|\rho-\sigma\|_1$ by diagonalizing $\rho-\sigma$.

**(b)** Compute the Uhlmann fidelity $F(\rho,\sigma) = \bigl[\mathrm{Tr}\sqrt{\sqrt{\rho}\,\sigma\,\sqrt{\rho}}\bigr]^2$; for two pure states it reduces to $|\langle 0|\psi\rangle|^2$.

**(c)** Verify the inequalities $1-\sqrt{F} \leq T \leq \sqrt{1-F}$, and determine for which $\alpha$ the upper bound is saturated.

**(d)** Using the Helstrom interpretation, state the optimal single-shot probability of telling the two states apart at $\alpha = \pi/2$.

## Solution

### Part (a): Trace distance

Write $c = \cos(\alpha/2)$ and $s = \sin(\alpha/2)$.  In the computational basis,

$$
\rho = \begin{pmatrix} 1 & 0 \\ 0 & 0\end{pmatrix},
\qquad
\sigma = \begin{pmatrix} c^2 & cs \\ cs & s^2 \end{pmatrix},
\qquad
\rho - \sigma = \begin{pmatrix} s^2 & -cs \\ -cs & -s^2 \end{pmatrix},
$$

using $1 - c^2 = s^2$ in the upper left entry.  The difference is Hermitian and traceless (both states have unit trace), so it is supported on the two-dimensional span of $|0\rangle, |1\rangle$ with eigenvalues $\pm\lambda$, and $\lambda$ follows from the determinant alone,

$$
\det(\rho-\sigma) = -s^4 - c^2 s^2 = -s^2\,(s^2+c^2) = -s^2 = -\lambda^2 ,
$$

so $\lambda = |s|$.  The trace norm of a Hermitian operator is the sum of the absolute values of its eigenvalues, hence

$$
\boxed{T(\rho,\sigma) = \tfrac{1}{2}\bigl(|\lambda| + |-\lambda|\bigr) = \lambda = |\sin(\alpha/2)| .}
$$

**The Bloch route.**  The same answer follows without diagonalizing anything.  For any two qubit states with Bloch vectors $\vec{r}_1, \vec{r}_2$, the difference $\rho_1 - \rho_2 = \frac{1}{2}\Delta\vec{r}\cdot\vec{\sigma}$ squares to $\frac{1}{4}|\Delta\vec{r}|^2\,\mathbb{I}$, so its eigenvalues are $\pm\frac{1}{2}|\Delta\vec{r}|$ and

$$
T(\rho_1,\rho_2) = \tfrac{1}{2}|\vec{r}_1 - \vec{r}_2| .
$$

Here $\vec{r}_\rho = (0,0,1)$ and $\vec{r}_\sigma = (\sin\alpha, 0, \cos\alpha)$, giving

$$
|\vec{r}_\rho-\vec{r}_\sigma|^2 = \sin^2\alpha + (1-\cos\alpha)^2 = 2(1-\cos\alpha) = 4\sin^2(\alpha/2),
$$

and therefore $T = |\sin(\alpha/2)|$ again.  Geometrically, the trace distance between two pure qubit states is half the chord joining them on the Bloch sphere, while $\alpha$ is the angle they subtend at the centre.

### Part (b): Uhlmann fidelity

Since $\rho = |0\rangle\langle 0|$ is a projector, $\sqrt{\rho} = \rho$, and the sandwich collapses immediately:

$$
\sqrt{\rho}\,\sigma\,\sqrt{\rho} = |0\rangle\langle 0|\psi\rangle\langle\psi|0\rangle\langle 0| = |\langle 0|\psi\rangle|^2\;|0\rangle\langle 0| = c^2\,|0\rangle\langle 0| .
$$

A non-negative multiple of a projector has square root $\sqrt{c^2}\,|0\rangle\langle 0| = |c|\,|0\rangle\langle 0|$, whose trace is $|c|$.  Squaring,

$$
\boxed{F(\rho,\sigma) = |\langle 0|\psi\rangle|^2 = \cos^2(\alpha/2).}
$$

This reproduces the general pure-state reduction of the Uhlmann formula, the squared overlap $F(|\psi_1\rangle\langle\psi_1|, |\psi_2\rangle\langle\psi_2|) = |\langle\psi_1|\psi_2\rangle|^2$.

### Part (c): Relating $T$ and $F$

With $T = |\sin(\alpha/2)|$ from part (a) and $\sqrt{F} = |\cos(\alpha/2)|$ from part (b), the upper bound evaluates to

$$
\sqrt{1-F} = \sqrt{1-\cos^2(\alpha/2)} = |\sin(\alpha/2)| = T .
$$

The upper bound is an equality at **every** $\alpha$.  This is not a coincidence of the parametrization.  The upper bound is saturated by any pair of pure states, and $\rho$ and $\sigma$ are pure for all $\alpha$.  A pair of states can therefore have high fidelity and still sit as far apart as the fidelity permits.

The lower bound reads

$$
1 - |\cos(\alpha/2)| \leq |\sin(\alpha/2)| ,
$$

which holds for all $\alpha$.  To locate the equality, put $u = |\cos(\alpha/2)|$ and $v = |\sin(\alpha/2)|$, so that $u,v \geq 0$ and $u^2 + v^2 = 1$.  Squaring $1-u = v$ gives $1 - 2u + u^2 = 1 - u^2$, i.e. $2u(u-1) = 0$.  Equality therefore requires $u = 0$ or $u = 1$, which on the principal range $\alpha \in [0,\pi]$ means $\alpha = \pi$ (orthogonal states, $T = 1$, $F = 0$) or $\alpha = 0$ (identical states, $T = 0$, $F = 1$).  In between the lower bound is strict, and the gap $T - (1-\sqrt{F}) = \sin(\alpha/2) + \cos(\alpha/2) - 1$ is largest at $\alpha = \pi/2$, where it equals $\sqrt{2}-1 \approx 0.414$.

### Part (d): Optimal discrimination at $\alpha = \pi/2$

At $\alpha = \pi/2$ the second state is $|\psi\rangle = (|0\rangle + |1\rangle)/\sqrt{2} = |+\rangle$, whose Bloch vector $(1,0,0)$ is orthogonal to $\vec{r}_\rho = (0,0,1)$.  The two states are as far apart as two pure states can be while remaining non-orthogonal in Hilbert space, since $\langle 0|+\rangle = 1/\sqrt{2} \neq 0$.  Part (a) gives

$$
T = \sin(\pi/4) = \frac{1}{\sqrt{2}} \approx 0.7071 ,
$$

and the Helstrom bound turns this into the optimal single-shot success probability for deciding which of the two equally likely states was prepared,

$$
\boxed{p_{\mathrm{correct}} = \tfrac{1}{2}\bigl(1 + T\bigr) = \tfrac{1}{2}\Bigl(1 + \tfrac{1}{\sqrt{2}}\Bigr) \approx 0.854 .}
$$

The measurement achieving this projects onto the positive and negative eigenspaces of $\rho - \sigma$, with the guess $\rho$ on the positive outcome and $\sigma$ on the negative one.  No POVM whatsoever does better, which is what the last code cell checks by brute force.  Chance alone would give $1/2$, and perfect discrimination would need $T = 1$, i.e. orthogonal supports.

## Symbolic Verification (SymPy)

```python
import sympy as sp

alpha = sp.Symbol('alpha', real=True)
c, s = sp.cos(alpha / 2), sp.sin(alpha / 2)

ket0 = sp.Matrix([1, 0])
psi = sp.Matrix([c, s])
rho = ket0 * ket0.T
sigma = psi * psi.T
diff = rho - sigma

# Part (a): rho - sigma is traceless, so its eigenvalues are ±sqrt(-det)
print("rho - sigma =", diff.tolist())
print("trace       =", sp.simplify(diff.trace()))
det = sp.simplify(diff.det())
print("det         =", det, "  (analytic: -sin(alpha/2)**2)")
assert sp.simplify(det + s**2) == 0

evals = sorted(diff.eigenvals(), key=sp.default_sort_key)
print("eigenvalues =", [sp.simplify(e) for e in evals])

T_sym = sp.sqrt(-det)
print("T(rho,sigma) =", T_sym, "  (analytic: |sin(alpha/2)|)")
assert sp.simplify(T_sym - sp.Abs(s)) == 0
```

```python
# Part (b): sqrt(rho) = rho for a projector, so the Uhlmann sandwich collapses
sandwich = sp.simplify(rho * sigma * rho)
print("sqrt(rho) sigma sqrt(rho) =", sandwich.tolist(), " = cos(alpha/2)**2 |0><0|")
assert sp.simplify(sandwich - c**2 * rho) == sp.zeros(2)

F_sym = sp.simplify((ket0.T * psi)[0, 0] ** 2)
print("F(rho,sigma)              =", F_sym, "  (analytic: cos(alpha/2)**2)")
assert sp.simplify(F_sym - c**2) == 0
```

```python
# Part (c): the upper bound is an identity for these pure states
upper = sp.sqrt(1 - F_sym)
print("sqrt(1 - F) =", sp.simplify(upper), "  and  T =", sp.Abs(s))
assert sp.simplify(upper - sp.Abs(s)) == 0
print("upper bound saturated for every alpha ✓")

# Lower-bound gap on 0 <= alpha <= pi, where |sin| = sin and |cos| = cos
a = sp.Symbol('a', nonnegative=True)
gap = sp.sin(a / 2) + sp.cos(a / 2) - 1          # = T - (1 - sqrt(F))
crit = sp.solveset(sp.diff(gap, a), a, domain=sp.Interval(0, sp.pi))
print("gap stationary at alpha =", crit, " -> gap =", sp.simplify(gap.subs(a, sp.pi / 2)))
print("gap zeros at alpha =", sp.solveset(sp.Eq(gap, 0), a, domain=sp.Interval(0, sp.pi)))
```

## Numerical Verification (NumPy / SciPy)

```python
import numpy as np
from scipy.linalg import sqrtm

sx = np.array([[0, 1], [1, 0]], dtype=complex)
sy = np.array([[0, -1j], [1j, 0]])
sz = np.array([[1, 0], [0, -1]], dtype=complex)
I2 = np.eye(2, dtype=complex)


def trace_distance(a, b):
    """T = (1/2) * sum of |eigenvalues of (a - b)|, valid for Hermitian a, b."""
    return 0.5 * np.sum(np.abs(np.linalg.eigvalsh(a - b)))


def uhlmann_fidelity(a, b):
    """F = [Tr sqrt(sqrt(a) b sqrt(a))]^2, computed with dense matrix square roots."""
    sa = sqrtm(a)
    return float(np.real(np.trace(sqrtm(sa @ b @ sa)))) ** 2


def state(alpha):
    psi = np.array([np.cos(alpha / 2), np.sin(alpha / 2)], dtype=complex)
    return np.outer(psi, psi.conj())


rho = state(0.0)  # |0><0|

print("alpha/pi     T (num)   |sin(a/2)|      F (num)  cos^2(a/2)")
for frac in [0.0, 0.25, 0.5, 0.75, 1.0]:
    alpha = frac * np.pi
    sigma = state(alpha)
    T_num, F_num = trace_distance(rho, sigma), uhlmann_fidelity(rho, sigma)
    T_an, F_an = abs(np.sin(alpha / 2)), np.cos(alpha / 2) ** 2
    print(f"{frac:8.2f}  {T_num:9.6f}  {T_an:9.6f}   {F_num:10.6f}  {F_an:10.6f}")
    assert np.isclose(T_num, T_an, atol=1e-8)
    assert np.isclose(F_num, F_an, atol=1e-8)

print("\nParts (a) and (b) confirmed on the grid. ✓")
```

```python
# Part (c): the two inequalities across the whole principal range
alphas = np.linspace(0.0, np.pi, 401)
T = np.abs(np.sin(alphas / 2))
F = np.cos(alphas / 2) ** 2
lower_b, upper_b = 1 - np.sqrt(F), np.sqrt(1 - F)

assert np.all(lower_b <= T + 1e-12)
assert np.all(T <= upper_b + 1e-12)
print("1 - sqrt(F) <= T <= sqrt(1 - F) holds at all 401 grid points ✓")
print(f"max |T - sqrt(1-F)| = {np.max(np.abs(T - upper_b)):.2e}  (upper bound saturated everywhere)")

gap = T - lower_b
i = int(np.argmax(gap))
print(f"largest lower-bound gap = {gap[i]:.6f} at alpha = {alphas[i]/np.pi:.3f} pi"
      f"   (analytic: {np.sqrt(2)-1:.6f} at 0.500 pi)")
assert np.isclose(gap[i], np.sqrt(2) - 1, atol=1e-4)
tight = alphas[np.isclose(gap, 0.0, atol=1e-9)] / np.pi
print(f"lower bound tight at alpha/pi = {np.round(tight, 6)}   (analytic: 0 and 1)")
```

```python
# Part (d): the Helstrom bound at alpha = pi/2
alpha = np.pi / 2
sigma = state(alpha)

T = trace_distance(rho, sigma)
p_opt = 0.5 * (1 + T)
print(f"T(rho, sigma)       = {T:.6f}   (analytic 1/sqrt(2) = {1/np.sqrt(2):.6f})")
print(f"p_correct = (1+T)/2 = {p_opt:.6f}   (analytic {0.5*(1+1/np.sqrt(2)):.6f} ~ 0.854)")
assert np.isclose(T, 1 / np.sqrt(2))
assert np.isclose(p_opt, 0.5 * (1 + 1 / np.sqrt(2)))

# The optimal POVM: project onto the positive eigenspace of rho - sigma
w, v = np.linalg.eigh(rho - sigma)
P_pos = sum(np.outer(v[:, k], v[:, k].conj()) for k in range(2) if w[k] > 0)
P_neg = I2 - P_pos
p_helstrom = 0.5 * np.real(np.trace(P_pos @ rho)) + 0.5 * np.real(np.trace(P_neg @ sigma))
print(f"explicit Helstrom measurement: {p_helstrom:.6f}")
assert np.isclose(p_helstrom, p_opt)

# Brute force: no other two-outcome qubit measurement beats it
rng = np.random.default_rng(2024)
best = 0.0
for _ in range(50_000):
    n = rng.normal(size=3)
    n /= np.linalg.norm(n)
    E = 0.5 * (I2 + n[0] * sx + n[1] * sy + n[2] * sz)
    p = 0.5 * np.real(np.trace(E @ rho)) + 0.5 * np.real(np.trace((I2 - E) @ sigma))
    best = max(best, p)
print(f"best of 50000 random projective measurements: {best:.6f} <= {p_opt:.6f}")
assert best <= p_opt + 1e-9

print("\nAll assertions passed. ✓")
```
