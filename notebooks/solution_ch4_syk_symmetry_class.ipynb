# Exercise 4.4 — SYK symmetry classes via Jordan–Wigner

**Chapter 4: Scrambling Dynamics** &nbsp;|&nbsp; *Exercises*

## Motivation

The large-$N$ solution of the **SYK model** describes disorder-averaged correlators, and it says nothing about the discrete structure of any individual spectrum.  At finite $N_M$ the spectrum carries an exact piece of physics that the saddle point cannot see: the Hamiltonian possesses an antiunitary particle–hole symmetry whose square depends only on $N_M \bmod 8$, so the model realizes **all three** Wigner–Dyson classes in a period-eight cycle.

The Jordan–Wigner transformation makes this checkable.  It turns each Majorana into a Pauli string, the four-Majorana couplings into weighted Pauli strings, and the whole Hamiltonian into a $2^L \times 2^L$ matrix that can be diagonalized directly.  Nothing beyond the Pauli algebra of Chapter 1 is needed, and the level statistics of Chapter 2 then read off the class.

## Preliminaries / Toolbox

Before diving into the solution, recall the following concepts:

**1. Majorana operators:** $N_M$ Hermitian operators obeying the Clifford algebra $\{\hat\chi_a, \hat\chi_b\} = 2\delta_{ab}$.  Each squares to the identity, $\hat\chi_a^2 = \hat I$, and distinct ones anticommute.

**2. Jordan–Wigner map:** On $L = N_M/2$ qubits,
$$\hat\chi_{2k-1} = \Bigl(\textstyle\prod_{j<k}\hat\sigma^z_j\Bigr)\hat\sigma^x_k, \qquad \hat\chi_{2k} = \Bigl(\textstyle\prod_{j<k}\hat\sigma^z_j\Bigr)\hat\sigma^y_k .$$
The $\hat\sigma^z$ tail (the Jordan–Wigner string) records the fermionic sign.

**3. Fermion parity:** $\hat P = \prod_k \hat\sigma^z_k$ commutes with the four-Majorana Hamiltonian, so the spectrum splits into two sectors.  Mixing sectors would superpose two independent spectra and destroy level repulsion, so statistics must be taken within one sector.

**4. Gap ratio (Sec. 2.4):** $r_n = \min(s_n, s_{n+1})/\max(s_n, s_{n+1})$ with $s_n = E_{n+1}-E_n$, a dimensionless diagnostic needing no unfolding.  Its ensemble average takes the reference values
$$\langle r\rangle = 0.5307\ (\mathrm{GOE}), \quad 0.5996\ (\mathrm{GUE}), \quad 0.6744\ (\mathrm{GSE}), \quad 0.3863\ (\mathrm{Poisson}).$$

**5. Kramers degeneracy:** In the GSE class every level is exactly twofold degenerate.  These doublets are not spacings and must be collapsed to single levels before forming $s_n$, otherwise half the gaps vanish and $\langle r\rangle$ is meaningless.

## Exercise Statement

Using the Jordan–Wigner map, build the Majorana SYK Hamiltonian $\hat H=\sum_{i<j<k<l}J_{ijkl}\,\hat\chi_i\hat\chi_j\hat\chi_k\hat\chi_l$ (with $J_{ijkl}$ i.i.d. Gaussian) as a $2^L\times 2^L$ matrix on $L=N_M/2$ qubits.

**(a)** Verify that the Jordan–Wigner operators satisfy $\{\hat\chi_a,\hat\chi_b\}=2\delta_{ab}$.

**(b)** For $N_M = 8, 10, 12, 14$, project onto a fixed fermion-parity sector ($\hat P=\prod_k\hat\sigma^z_k=+1$), diagonalize, and compute the mean gap ratio $\langle r\rangle$.  For $N_M=12$ remove the twofold Kramers degeneracy before forming spacings.

**(c)** Identify the Wigner–Dyson class in each case and confirm the period-eight pattern GOE, GUE, GSE, GUE for $N_M\bmod 8 = 0,2,4,6$.

## Solution

### Part (a): the Majorana algebra

Write $\hat\chi_{2k-1}=Z_1\cdots Z_{k-1}X_k$ and $\hat\chi_{2k}=Z_1\cdots Z_{k-1}Y_k$, and take $\hat\chi_a$, $\hat\chi_b$ with $a \neq b$.

**Same qubit** ($a=2k-1$, $b=2k$).  The two strings carry an identical tail $Z_1\cdots Z_{k-1}$, which squares away, and differ only on site $k$, where $\{X_k, Y_k\}=0$.  The pair anticommutes.

**Different qubits** ($a$ on site $k$, $b$ on site $k' > k$).  The longer string reaches site $k$ with one extra $\hat\sigma^z_k$, which the shorter string does not have.  Commuting the two operators past each other therefore produces exactly one sign flip, from $\{\hat\sigma^z_k, \hat\sigma^x_k\} = \{\hat\sigma^z_k, \hat\sigma^y_k\} = 0$.  On every site strictly below $k$ both strings carry $\hat\sigma^z$ and commute; on sites between $k$ and $k'$ only the longer string acts, and on site $k'$ only the longer string's $\hat\sigma^x$ or $\hat\sigma^y$ acts against the shorter string's identity.  A single anticommuting site gives $\hat\chi_a\hat\chi_b = -\hat\chi_b\hat\chi_a$.

**Squares.**  Every factor squares to the identity, so $\hat\chi_a^2 = \hat I$.

Together these give
$$\boxed{\{\hat\chi_a,\hat\chi_b\}=2\delta_{ab}\,\hat I.}$$

The tails cancel pairwise inside a four-Majorana term, leaving a single Pauli string.  Two instances quoted in the text,
$$\hat\chi_1\hat\chi_2\hat\chi_3\hat\chi_4 = -\,\hat\sigma^z_1\hat\sigma^z_2, \qquad \hat\chi_1\hat\chi_3\hat\chi_5\hat\chi_7 = -\,\hat\sigma^y_1\hat\sigma^x_2\hat\sigma^y_3\hat\sigma^x_4,$$
are checked symbolically below.

### Part (b): gap ratio by exact diagonalization

The construction is mechanical.  Build the $N_M$ Majorana matrices, form $\hat H=\sum_{i<j<k<l}J_{ijkl}\,\hat\chi_i\hat\chi_j\hat\chi_k\hat\chi_l$ over all $\binom{N_M}{4}$ quartets with $J_{ijkl}\sim\mathcal{N}(0,1)$, restrict to the $\hat P = +1$ sector, and diagonalize.

Three points control the quality of the answer:

- **Parity projection.** $\hat P$ is diagonal in the computational basis, so the sector is selected by keeping the rows and columns where $\prod_k \hat\sigma^z_k = +1$.  The sector has dimension $2^{L-1}$.
- **Kramers doublets.** At $N_M=12$ every level appears twice.  Collapsing degenerate levels to one representative before differencing is what turns the raw spectrum into the GSE spacing sequence.  The same collapse is harmless in the other cases, where no exact degeneracy occurs.
- **Spectral bulk.** The density of states varies strongly near the edges, and while $r_n$ needs no unfolding it is still cleanest in the middle of the band.  Restricting to the central $60\%$ of spacings and averaging over disorder realizations gives a stable estimate.

### Part (c): the period-eight pattern

The class is fixed by an antiunitary particle–hole operator $\hat A = \hat{\mathcal{P}}\hat K$, built from complex conjugation $\hat K$ and a product of Majoranas $\hat{\mathcal{P}}$.  Whether $\hat A^2 = +1$, $\hat A^2 = -1$, or $\hat A$ fails to preserve the parity sector depends only on $N_M \bmod 8$, which is the Bott periodicity of the real Clifford algebras.  An antiunitary symmetry squaring to $+1$ makes the sector Hamiltonian real in a suitable basis (GOE); one squaring to $-1$ forces Kramers doublets and quaternionic structure (GSE); no antiunitary symmetry within the sector leaves a generic complex Hermitian matrix (GUE).

Reading off the numbers computed below,

| $N_M$ | $N_M \bmod 8$ | $\langle r\rangle$ (SYK, ED) | class | $\langle r\rangle$ (RMT) |
|---|---|---|---|---|
| 8  | 0 | $\approx 0.53$ | GOE | 0.531 |
| 10 | 2 | $\approx 0.60$ | GUE | 0.600 |
| 12 | 4 | $\approx 0.68$ | GSE | 0.674 |
| 14 | 6 | $\approx 0.60$ | GUE | 0.600 |

The all-to-all SYK model thus realizes all three Wigner–Dyson ensembles of Chapter 2, selected purely by the Majorana count modulo eight.

## Symbolic Verification (SymPy)

```python
import sympy as sp
from functools import reduce

I2 = sp.eye(2)
X = sp.Matrix([[0, 1], [1, 0]])
Y = sp.Matrix([[0, -sp.I], [sp.I, 0]])
Z = sp.Matrix([[1, 0], [0, -1]])

def kron(ops):
    return reduce(lambda a, b: sp.Matrix(sp.kronecker_product(a, b)), ops)

def majoranas_sym(L):
    '''N_M = 2L Jordan-Wigner Majoranas on L qubits, exact symbolic matrices.'''
    g = []
    for k in range(1, L + 1):
        tail = [Z] * (k - 1)
        g.append(kron(tail + [X] + [I2] * (L - k)))   # chi_{2k-1}
        g.append(kron(tail + [Y] + [I2] * (L - k)))   # chi_{2k}
    return g

# Part (a): {chi_a, chi_b} = 2 delta_ab, exactly
L = 3
g = majoranas_sym(L)
NM = 2 * L
ok = True
for a in range(NM):
    for b in range(NM):
        anti = sp.simplify(g[a] * g[b] + g[b] * g[a])
        target = 2 * sp.Integer(a == b) * sp.eye(2 ** L)
        ok = ok and (anti == target)
print(f'L={L} (N_M={NM}): {{chi_a, chi_b}} == 2 delta_ab  ->  {ok}')
assert ok
```

```python
# The two four-Majorana identities quoted in the text, checked exactly
L = 4
g = majoranas_sym(L)   # chi_1 .. chi_8 stored at indices 0..7

def pauli_string(ops):
    return kron([{'I': I2, 'X': X, 'Y': Y, 'Z': Z}[o] for o in ops])

lhs1 = g[0] * g[1] * g[2] * g[3]              # chi_1 chi_2 chi_3 chi_4
rhs1 = -pauli_string('ZZII')                   # -Z_1 Z_2
lhs2 = g[0] * g[2] * g[4] * g[6]              # chi_1 chi_3 chi_5 chi_7
rhs2 = -pauli_string('YXYX')                   # -Y_1 X_2 Y_3 X_4

print('chi_1 chi_2 chi_3 chi_4 == -Z_1 Z_2            ->', sp.simplify(lhs1 - rhs1) == sp.zeros(16, 16))
print('chi_1 chi_3 chi_5 chi_7 == -Y_1 X_2 Y_3 X_4    ->', sp.simplify(lhs2 - rhs2) == sp.zeros(16, 16))
assert sp.simplify(lhs1 - rhs1) == sp.zeros(16, 16)
assert sp.simplify(lhs2 - rhs2) == sp.zeros(16, 16)

# Each quartet is a single Pauli string: unitary, Hermitian up to sign, one entry per row
nz = sum(1 for e in lhs2 if e != 0)
print(f'\nnonzero entries of chi_1 chi_3 chi_5 chi_7: {nz} out of {16*16} (one per row, as a Pauli string)')
```

## Numerical Verification

```python
import numpy as np
from itertools import combinations
from functools import reduce

I2n = np.eye(2, dtype=complex)
Xn = np.array([[0, 1], [1, 0]], dtype=complex)
Yn = np.array([[0, -1j], [1j, 0]], dtype=complex)
Zn = np.array([[1, 0], [0, -1]], dtype=complex)

def kronn(ops):
    return reduce(np.kron, ops)

def majoranas(L):
    '''N_M = 2L Majorana operators on L qubits via Jordan-Wigner.'''
    g = []
    for k in range(1, L + 1):
        tail = [Zn] * (k - 1)
        g.append(kronn(tail + [Xn] + [I2n] * (L - k)))   # chi_{2k-1}
        g.append(kronn(tail + [Yn] + [I2n] * (L - k)))   # chi_{2k}
    return g

# Part (a) numerically, at the sizes actually used below
print('Part (a): Majorana anticommutation relations')
for L in (4, 5, 6, 7):
    g = majoranas(L)
    err = 0.0
    for a in range(2 * L):
        for b in range(2 * L):
            anti = g[a] @ g[b] + g[b] @ g[a]
            err = max(err, np.abs(anti - 2 * (a == b) * np.eye(2 ** L)).max())
    print(f'  L={L} (N_M={2*L:2d}): max |{{chi_a,chi_b}} - 2 delta_ab| = {err:.2e}')
    assert err < 1e-12
print('  Jordan-Wigner Majoranas verified.')
```

```python
def gap_ratios(L, nreal, seed=1):
    '''Gap ratios of the Majorana SYK spectrum in the P=+1 parity sector.'''
    rng = np.random.default_rng(seed)
    NM = 2 * L
    g = majoranas(L)

    # parity P = prod_k Z_k is diagonal; keep the P=+1 rows/columns
    pdiag = np.real(np.diag(kronn([Zn] * L)))
    idx = np.where(pdiag > 0)[0]

    # precompute the quartet Pauli strings once, already projected
    pair = {(i, j): g[i] @ g[j] for i in range(NM) for j in range(i + 1, NM)}
    combs = list(combinations(range(NM), 4))
    quart = np.array([(pair[(i, j)] @ pair[(k, l)])[np.ix_(idx, idx)]
                      for (i, j, k, l) in combs])

    out = []
    for _ in range(nreal):
        J = rng.normal(size=len(combs))
        H = np.tensordot(J, quart, axes=(0, 0))
        H = (H + H.conj().T) / 2
        E = np.sort(np.linalg.eigvalsh(H).real)

        # collapse exact (Kramers) degeneracies to single levels
        Eu = [E[0]]
        for e in E[1:]:
            if e - Eu[-1] > 1e-7 * max(1.0, abs(e)):
                Eu.append(e)
        n_levels = len(Eu)

        s = np.diff(np.array(Eu))
        n = len(s)
        s = s[int(0.2 * n):int(0.8 * n)]          # central bulk
        out.append(np.minimum(s[:-1], s[1:]) / np.maximum(s[:-1], s[1:]))
    return np.concatenate(out), n_levels, len(idx)

REF = {'GOE': 0.5307, 'GUE': 0.5996, 'GSE': 0.6744}
POISSON = 2 * np.log(2) - 1
EXPECTED = {8: 'GOE', 10: 'GUE', 12: 'GSE', 14: 'GUE'}
NREAL = {4: 200, 5: 200, 6: 100, 7: 20}

print('Reference:  GOE 0.5307   GUE 0.5996   GSE 0.6744   Poisson '
      f'{POISSON:.4f}\n')
print(f"{'N_M':>4} {'mod 8':>6} {'<r> (ED)':>10} {'class':>7} {'<r> (RMT)':>10} "
      f"{'levels':>8} {'sector':>7}")
print('-' * 60)

results = {}
for NM in (8, 10, 12, 14):
    L = NM // 2
    r, n_levels, dim = gap_ratios(L, NREAL[L])
    rbar = r.mean()
    cls = min(REF, key=lambda c: abs(REF[c] - rbar))
    results[NM] = rbar
    print(f'{NM:>4} {NM % 8:>6} {rbar:>10.4f} {cls:>7} {REF[cls]:>10.4f} '
          f'{n_levels:>8} {dim:>7}')
    assert cls == EXPECTED[NM], f'class mismatch at N_M={NM}'
    assert abs(rbar - REF[cls]) < 0.02, f'gap ratio off reference at N_M={NM}'

print('\nPeriod-8 pattern confirmed: N_M mod 8 = 0, 2, 4, 6  ->  GOE, GUE, GSE, GUE.')
```

```python
# The Kramers doublets are what distinguishes N_M = 12: check the level count directly
print('Exact degeneracy in the P=+1 sector (single disorder realization):\n')
print(f"{'N_M':>4} {'sector dim':>11} {'distinct levels':>16} {'ratio':>7}")
print('-' * 44)
for NM in (8, 10, 12, 14):
    L = NM // 2
    _, n_levels, dim = gap_ratios(L, 1)
    print(f'{NM:>4} {dim:>11} {n_levels:>16} {dim / n_levels:>7.1f}')

print('\nOnly N_M=12 shows the twofold Kramers degeneracy of the GSE class.')
```

## Visual Pedagogical Proof

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
fig, ax = plt.subplots(figsize=(8, 4.5))

NMs = [8, 10, 12, 14]
vals = [results[n] for n in NMs]
colors = {'GOE': 'royalblue', 'GUE': 'seagreen', 'GSE': 'crimson'}
bar_colors = [colors[EXPECTED[n]] for n in NMs]

ax.bar([str(n) for n in NMs], vals, color=bar_colors, alpha=0.8, width=0.55,
       zorder=3, label='SYK (exact diagonalization)')

for name, val in REF.items():
    ax.axhline(val, color=colors[name], linestyle='--', linewidth=1.5, zorder=2)
    ax.text(3.42, val, f' {name} {val:.3f}', va='center', fontsize=9,
            color=colors[name])
ax.axhline(POISSON, color='gray', linestyle=':', linewidth=1.5, zorder=2)
ax.text(3.42, POISSON, f' Poisson {POISSON:.3f}', va='center', fontsize=9,
        color='gray')

for i, (n, v) in enumerate(zip(NMs, vals)):
    ax.text(i, v + 0.008, f'{v:.3f}', ha='center', fontsize=9)
    ax.text(i, 0.03, f'mod 8 = {n % 8}\n{EXPECTED[n]}', ha='center',
            fontsize=9, color='white', weight='bold')

ax.set_xlabel(r'Number of Majoranas $N_M$')
ax.set_ylabel(r'Mean gap ratio $\langle r \rangle$')
ax.set_title('SYK level statistics cycle with period eight in $N_M$')
ax.set_ylim(0, 0.78)
ax.set_xlim(-0.6, 4.3)
ax.grid(True, axis='y', ls='--', alpha=0.4, zorder=0)
ax.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
