## Motivation

The **quantum Fisher information** sets the best precision any measurement can extract about a parameter $\theta$ encoded unitarily by a generator $\hat{G}$.  For pure probes it reduces to a variance, $\mathcal{F}_Q = 4\,\mathrm{Var}[\hat{G}]$, so the question of metrological power becomes a question about how widely the probe state spreads over the eigenvalues of $\hat{G}$.

Two probe states bracket the interesting range.  Uncorrelated particles give a variance that is additive, hence $\mathcal{F}_Q = N$, the **standard quantum limit**.  A maximally entangled GHZ probe puts all its weight on the two extreme eigenvalues of $\hat{G}$, giving $\mathcal{F}_Q = N^2$, the **Heisenberg limit**.  The factor of $N$ between them is the metrological payoff of entanglement, and it is exactly the resource the scrambling protocols of this chapter aim to generate dynamically and then protect against particle loss.

## Preliminaries / Toolbox

Before diving into the solution, recall the following concepts:

**1. Quantum Fisher information for pure states:** For $\hat{\rho}_0 = \ket{\psi_0}\!\bra{\psi_0}$ and unitary encoding $e^{-i\theta\hat{G}}$,
$$\mathcal{F}_Q = 4\,\mathrm{Var}_{\psi_0}[\hat{G}] = 4\left(\bra{\psi_0}\hat{G}^2\ket{\psi_0} - \bra{\psi_0}\hat{G}\ket{\psi_0}^2\right).$$
A probe with large fluctuations in $\hat{G}$ moves faster under the encoding and is easier to read out.

**2. Quantum Cramér–Rao bound:** $\Delta\theta \geq 1/\sqrt{\nu\,\mathcal{F}_Q}$ for $\nu$ repetitions.  A gain of $N$ in $\mathcal{F}_Q$ is a gain of $\sqrt{N}$ in precision.

**3. The collective generator:** $\hat{G} = \frac{1}{2}\sum_{j=1}^N \hat{Z}_j$ has eigenvalues $(N-2k)/2$ for $k$ excited qubits, so its spectrum runs from $-N/2$ to $+N/2$.

**4. Expanding the variance:** Since $\hat{Z}_j^2 = \hat{I}$,
$$\hat{G}^2 = \frac{1}{4}\Bigl(\sum_j \hat{Z}_j\Bigr)^2 = \frac{1}{4}\Bigl(N\hat{I} + \sum_{j\neq k}\hat{Z}_j\hat{Z}_k\Bigr),$$
so the whole problem reduces to the two-point function $\langle \hat{Z}_j\hat{Z}_k\rangle$.  This one identity handles both probe states.

## Exercise Statement

For phase estimation with generator $\hat{G}=\tfrac12\sum_{j=1}^N\hat{Z}_j$, compute the quantum Fisher information $F_Q=4\,\mathrm{Var}(\hat{G})$ for

**(i)** the product state $\ket{+}^{\otimes N}$, and

**(ii)** the GHZ state $(\ket{0}^{\otimes N}+\ket{1}^{\otimes N})/\sqrt{2}$.

## Solution

For a pure state $F_Q = 4\left(\langle\hat{G}^2\rangle - \langle\hat{G}\rangle^2\right)$.  Both parts follow from the expansion
$$\hat{G}^2 = \frac{1}{4}\Bigl(N\hat{I} + \sum_{j\neq k}\hat{Z}_j\hat{Z}_k\Bigr),$$
which uses only $\hat{Z}_j^2 = \hat{I}$.  What separates the two probes is the correlator $\langle\hat{Z}_j\hat{Z}_k\rangle$ for $j \neq k$.

### Part (i): the product state and the standard quantum limit

On $\ket{+}^{\otimes N}$ each qubit is an independent eigenstate of $\hat{X}$, so
$$\langle\hat{Z}_j\rangle = 0, \qquad \mathrm{Var}(\hat{Z}_j) = \langle\hat{Z}_j^2\rangle - \langle\hat{Z}_j\rangle^2 = 1 .$$
The state factorizes, so for $j \neq k$ the correlator factorizes too, $\langle\hat{Z}_j\hat{Z}_k\rangle = \langle\hat{Z}_j\rangle\langle\hat{Z}_k\rangle = 0$.  The cross terms drop and
$$\langle\hat{G}\rangle = 0, \qquad \langle\hat{G}^2\rangle = \frac{1}{4}\bigl(N + 0\bigr) = \frac{N}{4},$$
equivalently $\mathrm{Var}(\hat{G}) = \tfrac14\sum_j \mathrm{Var}(\hat{Z}_j) = N/4$ by additivity over independent qubits.  Hence
$$\boxed{\mathcal{F}_Q^{(\mathrm{prod})} = 4\cdot\frac{N}{4} = N,}$$
the **standard quantum limit**, giving $\Delta\theta \geq 1/\sqrt{\nu N}$.  This is the shot-noise scaling of averaging over $N$ independent probes.

### Part (ii): the GHZ state and the Heisenberg limit

On $\ket{\mathrm{GHZ}} = (\ket{0}^{\otimes N}+\ket{1}^{\otimes N})/\sqrt{2}$ the generator is diagonal, and each branch is an eigenvector:
$$\hat{G}\ket{0}^{\otimes N} = +\frac{N}{2}\ket{0}^{\otimes N}, \qquad \hat{G}\ket{1}^{\otimes N} = -\frac{N}{2}\ket{1}^{\otimes N}.$$
So $\hat{G}$ takes the values $\pm N/2$ with probability $1/2$ each, giving
$$\langle\hat{G}\rangle = \frac{1}{2}\Bigl(\frac{N}{2}\Bigr) + \frac{1}{2}\Bigl(-\frac{N}{2}\Bigr) = 0, \qquad \langle\hat{G}^2\rangle = \frac{N^2}{4}.$$
Equivalently, in the two-point language every pair is perfectly correlated, $\langle\hat{Z}_j\hat{Z}_k\rangle = 1$ for all $j\neq k$ (both branches have all $\hat{Z}$ values equal), so $\langle\hat{G}^2\rangle = \tfrac14\bigl(N + N(N-1)\bigr) = N^2/4$.  Hence $\mathrm{Var}(\hat{G}) = N^2/4$ and
$$\boxed{\mathcal{F}_Q^{(\mathrm{GHZ})} = 4\cdot\frac{N^2}{4} = N^2,}$$
the **Heisenberg limit**, giving $\Delta\theta \geq 1/(\sqrt{\nu}\,N)$.

### The metrological gain

$$\frac{\mathcal{F}_Q^{\mathrm{GHZ}}}{\mathcal{F}_Q^{\mathrm{prod}}} = \frac{N^2}{N} = N .$$

Numerically, $\mathcal{F}_Q = N$ versus $N^2$ for $N = 2, 3, 4$ gives $2$ vs $4$, $3$ vs $9$, and $4$ vs $16$.  The point of the scrambling-metrology protocols of this chapter is to generate this $N^2$ sensitivity dynamically and to protect it against the loss of individual qubits, since the bare GHZ advantage collapses when even one qubit is traced out.

## Symbolic Verification (SymPy)

```python
import sympy as sp

N = sp.Symbol('N', positive=True, integer=True)

# G^2 = (1/4)(N + sum_{j != k} Z_j Z_k), with N(N-1) ordered pairs j != k
def FQ(two_point):
    '''F_Q = 4 Var(G) for <G> = 0, given the correlator <Z_j Z_k>, j != k.'''
    G2 = sp.Rational(1, 4) * (N + N * (N - 1) * two_point)
    return sp.simplify(4 * G2)

FQ_prod = FQ(0)    # product state: <Z_j Z_k> = <Z_j><Z_k> = 0
FQ_ghz = FQ(1)     # GHZ state:     <Z_j Z_k> = 1

print('Product |+>^N :  <Z_j Z_k> = 0  ->  F_Q =', FQ_prod)
print('GHZ          :  <Z_j Z_k> = 1  ->  F_Q =', FQ_ghz)
print('gain F_Q(GHZ)/F_Q(prod)         =', sp.simplify(FQ_ghz / FQ_prod))

assert FQ_prod == N
assert sp.simplify(FQ_ghz - N**2) == 0
assert sp.simplify(FQ_ghz / FQ_prod - N) == 0
```

```python
# Cross-check the product state via additivity over independent qubits,
# using the binomial distribution of the number of excited qubits.
from sympy.stats import Binomial, E, variance

print('Product state, explicit binomial average of G = (N - 2k)/2:\n')
for n in range(2, 7):
    k = Binomial('k', n, sp.Rational(1, 2))
    G = (n - 2 * k) / 2
    var_G = sp.simplify(variance(G))
    print(f'  N={n}: <G> = {sp.simplify(E(G))},  Var(G) = {var_G},  '
          f'F_Q = {sp.simplify(4*var_G)}   (analytic {n})')
    assert sp.simplify(4 * var_G - n) == 0

print('\nVar(G) = N/4 and F_Q = N, the standard quantum limit.')
```

## Numerical Verification

```python
import numpy as np

def G_diagonal(N):
    '''Diagonal of G = (1/2) sum_j Z_j in the computational basis.'''
    bits = (np.arange(2**N)[:, None] >> np.arange(N - 1, -1, -1)) & 1
    return 0.5 * np.sum(1 - 2 * bits, axis=1)

def FQ_pure(psi, gdiag):
    '''F_Q = 4 Var(G) for a pure state, with G diagonal.'''
    p = np.abs(psi)**2
    g1 = np.sum(p * gdiag)
    g2 = np.sum(p * gdiag**2)
    return 4 * (g2 - g1**2)

print('F_Q = 4 Var(G) for G = (1/2) sum_j Z_j\n')
print(f"{'N':>3} {'F_Q prod':>10} {'analytic N':>11} {'F_Q GHZ':>10} "
      f"{'analytic N^2':>13} {'gain':>6}")
print('-' * 60)
for N in range(2, 11):
    D = 2**N
    gdiag = G_diagonal(N)

    psi_prod = np.ones(D) / np.sqrt(D)              # |+>^N
    psi_ghz = np.zeros(D)                           # (|0..0> + |1..1>)/sqrt(2)
    psi_ghz[0] = psi_ghz[-1] = 1 / np.sqrt(2)

    f_prod = FQ_pure(psi_prod, gdiag)
    f_ghz = FQ_pure(psi_ghz, gdiag)
    print(f'{N:>3} {f_prod:>10.4f} {N:>11} {f_ghz:>10.4f} {N**2:>13} '
          f'{f_ghz/f_prod:>6.2f}')

    assert abs(f_prod - N) < 1e-10
    assert abs(f_ghz - N**2) < 1e-10
    assert abs(f_ghz / f_prod - N) < 1e-10

print('\nBoth match the analytic values exactly. Gain = N.')
```

```python
# The same numbers built from the full density matrix and the mixed-state QFI,
# F_Q = 2 sum_{k,l} (lam_k - lam_l)^2 / (lam_k + lam_l) |<k|G|l>|^2.
def FQ_mixed(rho, G, tol=1e-12):
    lam, V = np.linalg.eigh(rho)
    Gm = V.conj().T @ G @ V
    out = 0.0
    for k in range(len(lam)):
        for l in range(len(lam)):
            s = lam[k] + lam[l]
            if s > tol:
                out += 2 * (lam[k] - lam[l])**2 / s * abs(Gm[k, l])**2
    return out

print('Pure-state formula vs the general mixed-state QFI:\n')
print(f"{'N':>3} {'4 Var(G)':>10} {'mixed-state QFI':>17}")
print('-' * 33)
for N in (2, 3, 4, 5):
    D = 2**N
    gdiag = G_diagonal(N)
    G = np.diag(gdiag).astype(complex)
    psi = np.zeros(D, dtype=complex)
    psi[0] = psi[-1] = 1 / np.sqrt(2)
    rho = np.outer(psi, psi.conj())
    a, b = FQ_pure(psi, gdiag), FQ_mixed(rho, G)
    print(f'{N:>3} {a:>10.4f} {b:>17.4f}')
    assert abs(a - b) < 1e-8

# Fragility: trace out one qubit of the GHZ probe and the advantage is gone
print('\nGHZ after tracing out a single qubit (Sec. 6.5, fragility):\n')
print(f"{'N':>3} {'F_Q(GHZ)':>10} {'F_Q(N-1 qubits kept)':>22}")
print('-' * 37)
for N in (2, 3, 4, 5):
    D = 2**N
    psi = np.zeros(D, dtype=complex)
    psi[0] = psi[-1] = 1 / np.sqrt(2)
    rho = np.outer(psi, psi.conj()).reshape(2**(N-1), 2, 2**(N-1), 2)
    rho_A = np.trace(rho, axis1=1, axis2=3)          # discard the last qubit
    G_A = np.diag(G_diagonal(N - 1)).astype(complex)
    print(f'{N:>3} {float(N**2):>10.4f} {FQ_mixed(rho_A, G_A):>22.4f}')

print('\nLosing one qubit destroys the coherence carrying the phase, and')
print('the Heisenberg advantage drops to zero.')
```

## Visual Pedagogical Proof

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4))

# Left: F_Q scaling
Ns = np.arange(2, 21)
fq_prod, fq_ghz = [], []
for N in Ns:
    D = 2**N if N <= 14 else None
    if D is not None:
        gdiag = G_diagonal(N)
        psi_p = np.ones(D) / np.sqrt(D)
        psi_g = np.zeros(D); psi_g[0] = psi_g[-1] = 1 / np.sqrt(2)
        fq_prod.append(FQ_pure(psi_p, gdiag))
        fq_ghz.append(FQ_pure(psi_g, gdiag))
    else:
        fq_prod.append(np.nan); fq_ghz.append(np.nan)

ax1.loglog(Ns, Ns.astype(float), 'k--', lw=1.4, alpha=0.7, label=r'$N$ (SQL)')
ax1.loglog(Ns, Ns.astype(float)**2, 'k:', lw=1.4, alpha=0.7,
           label=r'$N^2$ (HL)')
ax1.scatter(Ns[:13], fq_prod[:13], color='royalblue', s=30, zorder=5,
            label=r'$|+\rangle^{\otimes N}$')
ax1.scatter(Ns[:13], fq_ghz[:13], color='crimson', s=30, zorder=5,
            label=r'$|\mathrm{GHZ}\rangle$')
ax1.set_xlabel(r'Number of qubits $N$')
ax1.set_ylabel(r'$\mathcal{F}_Q$')
ax1.set_title('Standard quantum limit versus Heisenberg limit')
ax1.legend()
ax1.grid(True, which='both', ls='--', alpha=0.4)

# Right: why. The distribution of G in each probe state.
N = 10
D = 2**N
gdiag = G_diagonal(N)
edges = np.arange(-N/2 - 0.5, N/2 + 1, 1)

psi_p = np.ones(D) / np.sqrt(D)
psi_g = np.zeros(D); psi_g[0] = psi_g[-1] = 1 / np.sqrt(2)

ax2.hist(gdiag, bins=edges, weights=np.abs(psi_p)**2, color='royalblue',
         alpha=0.75, label=r'$|+\rangle^{\otimes N}$: $\mathrm{Var}=N/4$')
ax2.hist(gdiag, bins=edges, weights=np.abs(psi_g)**2, color='crimson',
         alpha=0.75, label=r'$|\mathrm{GHZ}\rangle$: $\mathrm{Var}=N^2/4$')
ax2.set_xlabel(r'eigenvalue of $\hat{G} = \frac{1}{2}\sum_j \hat{Z}_j$')
ax2.set_ylabel('probability')
ax2.set_title(rf'Spread of $\hat{{G}}$ at $N={N}$ sets $\mathcal{{F}}_Q$')
ax2.legend()
ax2.grid(True, ls='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
