This notebook constructs the $6 \times 6$ Gram matrix of $S_3$, inverts it symbolically, and recovers the three third-moment Weingarten functions. Nothing is looked up: the Weingarten functions are produced by matrix inversion, which is what they are. The poles at $D = 1, 2$ are then traced to an explicit linear dependence among the six permutation operators.

## 1. Motivation

At $k=2$ the two Weingarten weights $1/(D^2-1)$ and $-1/[D(D^2-1)]$ are small enough to memorize, and it is easy to mistake them for tabulated constants. They are not. The Weingarten function is the inverse of the Gram matrix of permutation operators,

$$\mathrm{Wg}(\sigma^{-1}\tau, D) = (G^{-1})_{\sigma\tau}, \qquad G_{\sigma\tau} = \mathrm{Tr}\bigl(W_{\sigma^{-1}\tau}\bigr) = D^{\,\#\mathrm{cycles}(\sigma^{-1}\tau)} ,$$

and the inverse exists only because the permutation operators, while not orthogonal, are linearly independent. Going to $k=3$ makes both points visible. The matrix is now $6 \times 6$, the answers are no longer memorable, and the construction has to be run rather than recalled. The poles that appear at small $D$ then have a structural explanation, which is the real content of the exercise.

## 2. Exercise Statement

The third-moment Weingarten functions follow from inverting the Gram matrix, with no formula to memorize.

1. Using $G_{\sigma\tau} = D^{\#\mathrm{cycles}(\sigma^{-1}\tau)}$, describe the $6 \times 6$ Gram matrix of $S_3$: which entries are $D^3$, $D^2$, $D$?
2. Argue that $G^{-1}$ has only three distinct entries, equal to $\mathrm{Wg}(\mathrm{id})$, $\mathrm{Wg}(\text{transposition})$, $\mathrm{Wg}(\text{3-cycle})$, and confirm

$$
\mathrm{Wg}(\mathrm{id}) = \frac{D^2-2}{D(D^2-1)(D^2-4)}, \qquad
\mathrm{Wg}(\text{transp.}) = \frac{-1}{(D^2-1)(D^2-4)},
$$
$$
\mathrm{Wg}(\text{3-cycle}) = \frac{2}{D(D^2-1)(D^2-4)} .
$$

3. Explain the poles at $D = 1, 2$.

## 3. Analytical Derivation

### Part 1: the structure of $G$

The entry $G_{\sigma\tau}$ depends on $\sigma$ and $\tau$ only through the cycle type of $\sigma^{-1}\tau$, and $S_3$ has exactly three cycle types. Fix a row $\sigma$. As $\tau$ runs over the six elements of $S_3$, the product $\sigma^{-1}\tau$ runs over all of $S_3$ exactly once, so each row sees every cycle type with the multiplicity of that class:

| $\sigma^{-1}\tau$ | cycle type | $\#\mathrm{cycles}$ | $G_{\sigma\tau}$ | per row |
|---|---|---|---|---|
| identity | $1^3$ | 3 | $D^3$ | 1 |
| transposition | $2\,1$ | 2 | $D^2$ | 3 |
| 3-cycle | $3$ | 1 | $D$ | 2 |

So $G$ carries $D^3$ on the diagonal (the single element with $\sigma^{-1}\tau = \mathrm{id}$ is $\tau = \sigma$), $D^2$ in three off-diagonal positions of each row, and $D$ in the remaining two. Every row and column has the same census, since the class sizes of $S_3$ are $1, 3, 2$.

### Part 2: three distinct entries in $G^{-1}$

The counting above says that $G_{\sigma\tau} = g(\sigma^{-1}\tau)$ for the class function $g(\pi) = D^{\#\mathrm{cycles}(\pi)}$. A matrix of this form is a convolution operator on the group algebra, and the inverse of such a matrix has the same form, $(G^{-1})_{\sigma\tau} = h(\sigma^{-1}\tau)$ for some function $h$. Moreover $g$ is constant on conjugacy classes, and so is $h$. This is the statement that $\mathrm{Wg}$ depends only on the cycle type of its argument, and since $S_3$ has three classes, $G^{-1}$ has three distinct entries:

$$\mathrm{Wg}(\mathrm{id}), \qquad \mathrm{Wg}(\text{transposition}), \qquad \mathrm{Wg}(\text{3-cycle}).$$

The three values are fixed by $G G^{-1} = \mathbb{I}$, which, grouped by cycle type, is three linear equations in three unknowns. Solving them (or inverting $G$ directly, as below) gives the stated expressions. At $D = 3$ they are $7/120$, $-1/40$ and $1/60$.

### Part 3: the poles

The common denominator $(D^2-1)(D^2-4)$ vanishes at $D = 1$ and $D = 2$. Both are cases of $D < k = 3$. There the six permutation operators acting on $(\mathbb{C}^D)^{\otimes 3}$ are linearly dependent, so $G$, being their matrix of Hilbert-Schmidt overlaps, is singular and cannot be inverted. The inverse-Gram representation of the Haar average fails, even though the Haar integrals themselves remain perfectly finite. Nothing diverges physically; what breaks is the expansion in a set of operators that has ceased to be a basis.

## 4. Building the Gram Matrix

Permutations are stored as tuples, with `p[a]` the image of `a`. Composition, inversion and cycle counting are written out explicitly so that no part of the construction is hidden.

```python
import itertools
import sympy as sp
from sympy import Symbol, factor, simplify

D = Symbol('D', positive=True, integer=True)

perms = list(itertools.permutations(range(3)))     # the six elements of S_3

def compose(p, q):
    # (p o q)(a) = p(q(a))
    return tuple(p[q[a]] for a in range(3))

def inverse(p):
    r = [0] * 3
    for a, val in enumerate(p):
        r[val] = a
    return tuple(r)

def n_cycles(p):
    seen, count = set(), 0
    for a in range(3):
        if a not in seen:
            count += 1
            b = a
            while b not in seen:
                seen.add(b)
                b = p[b]
    return count

def cycle_type(p):
    return {3: 'identity', 2: 'transposition', 1: '3-cycle'}[n_cycles(p)]

print('The six elements of S_3 (as images of 0,1,2):')
for p in perms:
    print(f'  {p} : {n_cycles(p)} cycles -> {cycle_type(p)}')
```

```python
# G_{sigma,tau} = D^{#cycles(sigma^{-1} tau)}
G = sp.Matrix(6, 6, lambda a, b: D**n_cycles(compose(inverse(perms[a]), perms[b])))

print('Gram matrix of S_3:')
sp.pprint(G)

# Part 1: the census of each row
from collections import Counter
print()
print('Entries per row, by cycle type of sigma^{-1} tau:')
for a in range(1):
    hist = Counter(cycle_type(compose(inverse(perms[a]), perms[b])) for b in range(6))
    for name, power in [('identity', 3), ('transposition', 2), ('3-cycle', 1)]:
        print(f'  {name:14s} -> G = D^{power} : {hist[name]} per row')

# Every row and column must have the same census (the class sizes 1, 3, 2).
for a in range(6):
    hist = Counter(cycle_type(compose(inverse(perms[a]), perms[b])) for b in range(6))
    assert hist == Counter({'transposition': 3, '3-cycle': 2, 'identity': 1})
print()
print('Every row has the same census (class sizes of S_3 are 1, 3, 2).  PASS')
print('Book: D^3 on the diagonal, D^2 in three positions per row, D in the remaining two.  PASS')
```

## 5. Inverting the Gram Matrix

The Weingarten functions are now obtained by inversion, and compared against the expressions quoted in the exercise.

```python
G_inv = sp.simplify(G.inv())

# Written out in full the entries are unwieldy, and the point is their pattern,
# so display G^{-1} with one symbol standing for each cycle type.
a_s, b_s, c_s = sp.symbols('a b c')
label = {'identity': a_s, 'transposition': b_s, '3-cycle': c_s}
pattern = sp.Matrix(6, 6, lambda i, j: label[cycle_type(compose(inverse(perms[i]), perms[j]))])

print('Pattern of G^{-1}   [a = Wg(id), b = Wg(transposition), c = Wg(3-cycle)]:')
sp.pprint(pattern)
print()
print('The same shape as G itself: a on the diagonal, b in three places per row, c in two.')
print('The next cell checks that G^{-1} really is constant on each class, and')
print('extracts the three values.')
```

```python
# Part 2: collect the entries of G^{-1} by the cycle type of sigma^{-1} tau.
by_type = {}
for a in range(6):
    for b in range(6):
        t = cycle_type(compose(inverse(perms[a]), perms[b]))
        val = sp.simplify(G_inv[a, b])
        by_type.setdefault(t, set()).add(sp.nsimplify(val))

print('Distinct entries of G^{-1}, grouped by cycle type:')
for t in ['identity', 'transposition', '3-cycle']:
    vals = by_type[t]
    assert len(vals) == 1, f'{t} is not constant on its class'
    print(f'  Wg({t:14s}) = {factor(list(vals)[0])}')

all_vals = set().union(*by_type.values())
print()
print(f'Number of distinct entries in G^{{-1}} = {len(all_vals)}')
assert len(all_vals) == 3
print('G^{-1} is a class function, exactly three distinct values.  PASS')
```

```python
# Compare against the expressions quoted in the exercise.
Wg_id_book     = (D**2 - 2) / (D * (D**2 - 1) * (D**2 - 4))
Wg_transp_book = -1 / ((D**2 - 1) * (D**2 - 4))
Wg_3cyc_book   = 2 / (D * (D**2 - 1) * (D**2 - 4))

computed = {t: list(by_type[t])[0] for t in by_type}
book = {'identity': Wg_id_book, 'transposition': Wg_transp_book, '3-cycle': Wg_3cyc_book}

print(f'{"cycle type":>15} | {"computed (inverting G)":>40} | {"book":>40} | match')
print('-' * 115)
for t in ['identity', 'transposition', '3-cycle']:
    diff = sp.simplify(computed[t] - book[t])
    ok = (diff == 0)
    print(f'{t:>15} | {str(factor(computed[t])):>40} | {str(factor(book[t])):>40} | {"PASS" if ok else "FAIL"}')
    assert ok

print()
print('All three Weingarten functions reproduce the book expressions identically in D.  PASS')
```

```python
# The D = 3 values quoted in the solution.
print('At D = 3:')
for t, expected in [('identity', sp.Rational(7, 120)),
                    ('transposition', sp.Rational(-1, 40)),
                    ('3-cycle', sp.Rational(1, 60))]:
    val = sp.nsimplify(computed[t].subs(D, 3))
    print(f'  Wg({t:14s}) = {val}   (book: {expected})   {"PASS" if val == expected else "FAIL"}')
    assert val == expected

# Direct check that the inverse really is an inverse, symbolically in D.
print()
print('Verifying G * G^{-1} = I symbolically:')
prod = sp.simplify(G * G_inv)
assert prod == sp.eye(6)
print('  G * G^{-1} = I for general D.  PASS')
```

## 6. The Poles at $D = 1, 2$

The determinant shows where the inversion is legitimate, and building the permutation operators explicitly shows why it fails where it does.

```python
det_G = factor(G.det())
print(f'det G = {det_G}')
print()
print('Roots at positive integer D:')
for D_val in [1, 2, 3, 4]:
    d = det_G.subs(D, D_val)
    print(f'  D = {D_val}: det G = {d}   -> {"SINGULAR" if d == 0 else "invertible"}')

assert det_G.subs(D, 1) == 0 and det_G.subs(D, 2) == 0
assert det_G.subs(D, 3) != 0
print()
print('det G vanishes exactly at D = 1 and D = 2, i.e. for D < k = 3.')
print('These are the poles of the denominators (D^2-1)(D^2-4).  PASS')
```

```python
import numpy as np

def permutation_operator(p, D_val):
    # W_p acting on (C^D)^{ot 3}, permuting the three tensor factors.
    dim = D_val**3
    W = np.zeros((dim, dim))
    for idx in itertools.product(range(D_val), repeat=3):
        out = tuple(idx[p[a]] for a in range(3))
        row = sum(o * D_val**(2 - a) for a, o in enumerate(out))
        col = sum(i * D_val**(2 - a) for a, i in enumerate(idx))
        W[row, col] = 1
    return W

print('Rank of the span of the six permutation operators on (C^D)^{ot 3}:')
print()
print(f'{"D":>3} | {"rank":>4} | {"independent?":>13} | {"G invertible?":>14}')
print('-' * 45)
for D_val in [1, 2, 3, 4, 5]:
    vecs = np.array([permutation_operator(p, D_val).flatten() for p in perms])
    rank = np.linalg.matrix_rank(vecs, tol=1e-9)
    indep = (rank == 6)
    det_ok = (det_G.subs(D, D_val) != 0)
    print(f'{D_val:3d} | {rank:4d} | {"yes" if indep else "no":>13} | {"yes" if det_ok else "no":>14}')
    # The Gram matrix is invertible exactly when the operators are independent.
    assert indep == det_ok

print()
print('At D = 1 the six operators collapse to a single scalar (rank 1).')
print('At D = 2 they span only 5 dimensions, one linear dependence, so G is singular.')
print('From D = 3 = k onwards they are independent and G can be inverted.')
print('Linear independence and invertibility of G coincide exactly.  PASS')
```

The ranks $1, 5, 6, 6, 6$ are not accidental. By Schur-Weyl duality the span of the permutation operators is the commutant of $\hat U^{\otimes 3}$, whose dimension is

$$\sum_{\lambda \vdash 3,\ \ell(\lambda) \leq D} (f^\lambda)^2 ,$$

the sum running over partitions of 3 with at most $D$ rows, with $f^\lambda$ the number of standard Young tableaux of shape $\lambda$. Only when every partition of 3 fits, which requires $D \geq 3$, does the sum reach $\sum_{\lambda \vdash 3} (f^\lambda)^2 = 3! = 6$ and the permutations become a basis.

```python
import math

def partitions_of(n):
    out = []
    def rec(rem, mx, cur):
        if rem == 0:
            out.append(tuple(cur))
            return
        for k in range(min(rem, mx), 0, -1):
            rec(rem - k, k, cur + [k])
    rec(n, n, [])
    return out

def n_standard_tableaux(lam):
    # Hook length formula
    n = sum(lam)
    conj = [sum(1 for r in lam if r > j) for j in range(lam[0])]
    hooks = 1
    for i, r in enumerate(lam):
        for j in range(r):
            hooks *= (r - j) + (conj[j] - i) - 1
    return math.factorial(n) // hooks

print('Commutant dimension of U^{ot 3} from Schur-Weyl duality:')
print()
for D_val in [1, 2, 3, 4]:
    contributions = [(lam, n_standard_tableaux(lam)) for lam in partitions_of(3)
                     if len(lam) <= D_val]
    dim = sum(f**2 for _, f in contributions)
    detail = ' + '.join(f'{f}^2' for _, f in contributions)
    vecs = np.array([permutation_operator(p, D_val).flatten() for p in perms])
    rank = np.linalg.matrix_rank(vecs, tol=1e-9)
    print(f'  D={D_val}: partitions {[lam for lam, _ in contributions]}')
    print(f'        dim = {detail} = {dim}   (measured rank {rank})   '
          f'{"PASS" if dim == rank else "FAIL"}')
    assert dim == rank

print()
print('The rank sequence 1, 5, 6, 6 is the commutant dimension, and it saturates at 3! = 6')
print('once D >= k = 3.  PASS')
```

## 7. Physical Interpretation

The exercise is a worked demonstration that the Weingarten function is not a table but a matrix inverse. The Gram matrix counts loops, one factor of $D$ per cycle of $\sigma^{-1}\tau$, and the whole of $\mathrm{Wg}$ at $k=3$ follows from inverting that $6 \times 6$ array of loop counts. The class-function structure is what keeps the answer manageable: six permutations, but only three distinct Weingarten values, because the group cannot distinguish elements within a conjugacy class.

The hierarchy of the three values is worth reading off. At large $D$ the identity dominates at order $1/D^3$, a transposition enters at $-1/D^4$, and a 3-cycle at $+2/D^5$. Each transposition needed to build a permutation costs a factor of $1/D$, and the alternating signs are the Möbius function of the partition lattice. The leading coefficient of the 3-cycle is 2 rather than 1, which is the first place where the naive guess that these are all $\pm 1$ fails.

The poles carry the most physics. They are not a breakdown of Haar integration, since the moments of a Haar-random unitary are bounded for every $D$ and every $k$. They are the breakdown of a particular representation, the expansion of the Haar average in permutation operators. When $D < k$ those operators are linearly dependent, the rank calculation above puts the dependence at $D = 2$ explicitly (five independent operators, not six), and the dual basis that defines $\mathrm{Wg}$ ceases to exist. The condition $D \geq k$ that accompanies every Weingarten formula in this chapter is exactly the condition that the permutations remain a basis, and the same counting reappears in the frame potential, where $F^{(t)}_{\mathrm{Haar}} = t!$ holds only for $D \geq t$ and drops to the smaller commutant dimension below it.
