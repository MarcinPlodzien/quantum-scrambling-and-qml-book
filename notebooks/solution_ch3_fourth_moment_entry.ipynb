**Chapter 3: Haar Ensembles** &nbsp;|&nbsp; *Exercises*

This notebook evaluates two fourth-order moments of the entries of a Haar-random $\hat U \in U(D)$ using the $k=2$ Weingarten recipe. Every step of the index bookkeeping is carried out symbolically with SymPy, and the results are checked against Monte Carlo sampling and against the exact Dirichlet law of a single row.

## 1. Motivation

A single entry $U_{ij}$ of a Haar-random unitary is not an independent random variable. Each row is normalized,

$$\sum_{j} |U_{ij}|^2 = 1,$$

and that one constraint couples every entry of a row to every other. The two averages computed here are the smallest quantities that resolve the coupling:

- $\mathbb{E}\bigl[|U_{ij}|^4\bigr]$, the fourth moment of one entry,
- $\mathbb{E}\bigl[|U_{ij}|^2 |U_{ik}|^2\bigr]$ with $j \neq k$, the cross moment of two entries in the same row.

Each contains two factors of $U$ and two of $U^*$, so each is a $k=2$ Haar average and follows from the same index-pairing formula that gave $\mathbb{E}[|\mathrm{Tr}\,\hat U|^4] = 2$. The only thing that changes between the two is which Kronecker deltas survive.

## 2. Exercise Statement

Using the recipe of Sec. 3.2, evaluate two fourth-order moments of the entries of a Haar-random $\hat U \in U(D)$:

1. $\mathbb{E}\bigl[|U_{ij}|^4\bigr]$;
2. $\mathbb{E}\bigl[|U_{ij}|^2 |U_{ik}|^2\bigr]$ for $j \neq k$ (two entries in the same row).

## 3. Analytical Derivation

Both averages are governed by the $k=2$ index-pairing formula

$$
\mathbb{E}\bigl[U_{i_1 j_1} U_{i_2 j_2} U^*_{k_1 l_1} U^*_{k_2 l_2}\bigr]
= \sum_{\sigma, \tau \in S_2} \mathrm{Wg}(\sigma\tau^{-1}, D)\,
\prod_{a=1}^{2} \delta_{i_a,\, k_{\sigma(a)}}\; \delta_{j_a,\, l_{\tau(a)}},
$$

with $S_2 = \{\mathrm{id}, (12)\}$ and

$$
\mathrm{Wg}(\mathrm{id}, D) = \frac{1}{D^2-1}, \qquad
\mathrm{Wg}\bigl((12), D\bigr) = \frac{-1}{D(D^2-1)}.
$$

The permutation $\sigma$ pairs the row indices, $\tau$ pairs the column indices, and the weight depends only on the mismatch $\sigma\tau^{-1}$ between the two pairings. Which of the four $(\sigma, \tau)$ terms survives is decided entirely by the Kronecker deltas, and that is the only question to settle in each part.

### Part 1: $\mathbb{E}[|U_{ij}|^4]$

Write the moment with explicit indices,

$$|U_{ij}|^4 = U_{ij}\, U_{ij}\, U^*_{ij}\, U^*_{ij},$$

so that every row label equals $i$ and every column label equals $j$:

$$i_1 = i_2 = k_1 = k_2 = i, \qquad j_1 = j_2 = l_1 = l_2 = j.$$

Every delta then reads $\delta_{i,i}$ or $\delta_{j,j}$ and is satisfied no matter how $\sigma$ and $\tau$ are chosen. All four terms survive. Grouping them by their mismatch, the two pairs with $\sigma = \tau$ carry $\mathrm{Wg}(\mathrm{id})$ and the two with $\sigma \neq \tau$ carry $\mathrm{Wg}((12))$, giving $2[\mathrm{Wg}(\mathrm{id}) + \mathrm{Wg}((12))]$:

$$
\mathbb{E}\bigl[|U_{ij}|^4\bigr]
= \frac{2}{D^2-1} - \frac{2}{D(D^2-1)}
= \frac{2(D-1)}{D(D^2-1)}
= \boxed{\frac{2}{D(D+1)}}.
$$

### Part 2: $\mathbb{E}[|U_{ij}|^2 |U_{ik}|^2]$ with $j \neq k$

Now

$$|U_{ij}|^2 |U_{ik}|^2 = U_{ij}\, U_{ik}\, U^*_{ij}\, U^*_{ik},$$

so the row labels are still all equal to $i$, but the column labels split:

$$i_1 = i_2 = k_1 = k_2 = i, \qquad j_1 = l_1 = j, \quad j_2 = l_2 = k.$$

The row deltas are satisfied for both choices of $\sigma$, exactly as before. The column deltas are not. For $\tau = \mathrm{id}$ they read $\delta_{j,j}\, \delta_{k,k} = 1$, but for $\tau = (12)$ they read

$$\delta_{j_1,\, l_2}\, \delta_{j_2,\, l_1} = \delta_{jk}\, \delta_{kj} = 0,$$

since $j \neq k$. Only the direct column pairing $\tau = \mathrm{id}$ survives, and with it the weight reduces to $\mathrm{Wg}(\sigma\tau^{-1}) = \mathrm{Wg}(\sigma)$. Summing $\mathrm{Wg}(\sigma)$ over the two surviving $\sigma \in S_2$,

$$
\mathbb{E}\bigl[|U_{ij}|^2 |U_{ik}|^2\bigr]
= \mathrm{Wg}(\mathrm{id}) + \mathrm{Wg}\bigl((12)\bigr)
= \frac{1}{D^2-1} - \frac{1}{D(D^2-1)}
= \boxed{\frac{1}{D(D+1)}}.
$$

### Comparing the two

For $D = 2$ these are $1/3$ and $1/6$. The first is the larger, so entries in a row are negatively correlated, as the constraint $\sum_j |U_{ij}|^2 = 1$ requires.

## 4. Symbolic Verification (SymPy)

```python
import sympy as sp
from sympy import Symbol, Rational, cancel, factor

D = Symbol('D', positive=True, integer=True)

# k=2 Weingarten weights, indexed by the mismatch sigma*tau^{-1}
Wg_id = Rational(1) / (D**2 - 1)
Wg_12 = Rational(-1) / (D * (D**2 - 1))
print(f'Wg(id, D)   = {Wg_id}')
print(f'Wg((12), D) = {Wg_12}')
print()

# Part 1: all row and column labels equal, so every delta is satisfied
# and all four (sigma, tau) terms survive.
print('Part 1:  E[|U_ij|^4]   (all row and column labels equal)')
terms_1 = [('id',   'id',   'id',   Wg_id),
           ('(12)', '(12)', 'id',   Wg_id),
           ('id',   '(12)', '(12)', Wg_12),
           ('(12)', 'id',   '(12)', Wg_12)]
m4 = sp.Integer(0)
for sigma, tau, mismatch, wg in terms_1:
    m4 += wg
    print(f'  sigma={sigma:4s} tau={tau:4s} -> mismatch {mismatch:4s} : deltas all 1, Wg = {wg}')
m4 = factor(cancel(m4))
print(f'  sum over the four terms = {m4}')
print()

# Part 2: j != k kills every term with tau = (12); both sigma survive.
print('Part 2:  E[|U_ij|^2 |U_ik|^2],  j != k')
terms_2 = [('id',   'id', 'id',   Wg_id),
           ('(12)', 'id', '(12)', Wg_12)]
mx = sp.Integer(0)
for sigma, tau, mismatch, wg in terms_2:
    mx += wg
    print(f'  sigma={sigma:4s} tau={tau:4s} -> mismatch {mismatch:4s} : Wg = {wg}')
print('  sigma=id   tau=(12)            : column deltas give delta_{jk} = 0, term dropped')
print('  sigma=(12) tau=(12)            : column deltas give delta_{jk} = 0, term dropped')
mx = factor(cancel(mx))
print(f'  sum over the two surviving terms = {mx}')
print()

assert sp.simplify(m4 - 2 / (D * (D + 1))) == 0
assert sp.simplify(mx - 1 / (D * (D + 1))) == 0
print(f'E|U_ij|^4         = {m4}      (book: 2/(D(D+1)))   PASS')
print(f'E|U_ij|^2|U_ik|^2 = {mx}      (book: 1/(D(D+1)))   PASS')
print(f'At D=2: {m4.subs(D, 2)} and {mx.subs(D, 2)}   (book: 1/3 and 1/6)   PASS')
```

### Consistency with row normalization

The two moments are not independent. Squaring $\sum_j |U_{ij}|^2 = 1$ and averaging gives $D$ diagonal terms $\mathbb{E}|U_{ij}|^4$ and $D(D-1)$ off-diagonal terms $\mathbb{E}[|U_{ij}|^2|U_{ik}|^2]$, and the total must be exactly 1 for every $D$. This ties the two answers together and would expose a relative error in either.

```python
# E[(sum_j |U_ij|^2)^2] = D * E|U_ij|^4 + D(D-1) * E[|U_ij|^2 |U_ik|^2] = 1
sum_rule = sp.simplify(D * m4 + D * (D - 1) * mx)
print(f'D * E|U_ij|^4 + D(D-1) * E[|U_ij|^2|U_ik|^2] = {sum_rule}')
assert sum_rule == 1
print('Row normalization is satisfied identically in D.  PASS')
print()

# The negative correlation is what the constraint demands: with x_j = |U_ij|^2 and
# E[x_j] = 1/D, the connected correlator is strictly negative for every finite D.
cov = sp.factor(sp.simplify(mx - Rational(1, 1) / D**2))
print(f'Cov(|U_ij|^2, |U_ik|^2) = E[x_j x_k] - E[x_j]E[x_k] = {cov}')
print(f'At D=2: {cov.subs(D, 2)}  (negative, as the sum rule requires)')
```

### An independent route: the Dirichlet law of a row

The Weingarten calculus is not the only way to these numbers. The squared moduli of a single row of a Haar-random unitary, $x_j = |U_{ij}|^2$, are distributed uniformly on the probability simplex, which is the Dirichlet distribution with all parameters equal to 1. Its moment formula

$$\mathbb{E}\Bigl[\prod_j x_j^{a_j}\Bigr] = \frac{\Gamma(D)}{\Gamma\bigl(D + \sum_j a_j\bigr)} \prod_j \Gamma(1 + a_j)$$

reproduces both results with no Weingarten input at all, which makes it a genuine cross-check rather than a restatement.

```python
from sympy import gamma

def dirichlet_moment(exponents, dim):
    # E[prod_j x_j^{a_j}] for x ~ Dirichlet(1, ..., 1) on the (dim-1)-simplex
    num = gamma(dim) * sp.prod([gamma(1 + a) for a in exponents])
    return num / gamma(dim + sum(exponents))

d4 = sp.simplify(dirichlet_moment([2], D))        # E[x_j^2]   <-> E|U_ij|^4
dx = sp.simplify(dirichlet_moment([1, 1], D))     # E[x_j x_k] <-> E[|U_ij|^2|U_ik|^2]

print(f'Dirichlet  E[x_j^2]   = {factor(d4)}')
print(f'Dirichlet  E[x_j x_k] = {factor(dx)}')
print()
assert sp.simplify(d4 - m4) == 0
assert sp.simplify(dx - mx) == 0
print('Both agree with the Weingarten results identically in D.  PASS')
```

## 5. Numerical Verification (Monte Carlo)

```python
import numpy as np
from scipy.stats import unitary_group

rng = np.random.default_rng(42)
n_samples = 40000

print(f'Monte Carlo over {n_samples} Haar-random unitaries')
print()
print(f'{"D":>3} | {"E|U_ij|^4 (MC)":>16} {"analytic":>10} | '
      f'{"E|U_ij|^2|U_ik|^2 (MC)":>24} {"analytic":>10}')
print('-' * 76)

for D_val in [2, 3, 4, 8]:
    U = unitary_group.rvs(D_val, size=n_samples, random_state=rng)
    x_j = np.abs(U[:, 0, 0])**2      # |U_00|^2
    x_k = np.abs(U[:, 0, 1])**2      # |U_01|^2, same row, different column

    mc_4 = np.mean(x_j**2)
    mc_x = np.mean(x_j * x_k)
    an_4 = float(m4.subs(D, D_val))
    an_x = float(mx.subs(D, D_val))

    print(f'{D_val:3d} | {mc_4:16.5f} {an_4:10.5f} | {mc_x:24.5f} {an_x:10.5f}')
    assert abs(mc_4 - an_4) < 0.03 * an_4 + 1e-3
    assert abs(mc_x - an_x) < 0.03 * an_x + 1e-3

print()
print('Monte Carlo confirms both moments for every D tested.  PASS')
```

```python
# The D=2 numbers quoted in the text, to higher precision, plus the sum rule
# checked directly on the samples rather than on the formulas.
U = unitary_group.rvs(2, size=200000, random_state=rng)
row = np.abs(U[:, 0, :])**2                      # both entries of row 0

mc_4 = np.mean(row[:, 0]**2)
mc_x = np.mean(row[:, 0] * row[:, 1])
print(f'D=2:  E|U_ij|^4         = {mc_4:.5f}   (analytic 1/3 = {1/3:.5f})')
print(f'D=2:  E[|U_ij|^2|U_ik|^2] = {mc_x:.5f}   (analytic 1/6 = {1/6:.5f})')
print(f'D=2:  ratio                = {mc_4/mc_x:.4f}   (analytic 2)')
print()
print(f'Sampled E[(sum_j |U_ij|^2)^2] = {np.mean(row.sum(axis=1)**2):.10f}  (exactly 1)')
```

## 6. Physical Interpretation

The two moments differ by exactly a factor of 2, and the factor counts pairings rather than anything dimensional. In part 1 both index pairings are free and all four $(\sigma, \tau)$ terms contribute; in part 2 the distinct column labels freeze $\tau$ to the identity and only half the terms remain. The Weingarten weights are identical in the two cases, so the entire difference is combinatorial.

The inequality $\mathbb{E}|U_{ij}|^4 > \mathbb{E}[|U_{ij}|^2|U_{ik}|^2]$ has a direct physical reading. If a row of $\hat U$ had independent entries, the two averages would factorize and the cross moment would equal $(1/D)^2 = 1/D^2$, whereas the true value $1/(D(D+1))$ is smaller. Weight that lands on one entry of a row is weight denied to the others, since the total is pinned to 1, and the connected correlator computed above is negative for every finite $D$. Both moments scale as $1/D^2$ at large $D$, and the correlation they encode is a $1/D$ effect that disappears in the same limit where the entries become effectively independent complex Gaussians.

This is the entry-level counterpart of the trace calculation of Exercise 3.3. There the Weingarten sum produced the exactly dimension-independent answer $\mathbb{E}|\mathrm{Tr}\,\hat U|^4 = 2$; here the same four terms, filtered by different deltas, produce answers that retain their $D$ dependence. The finite-$D$ corrections that cancel for the trace are precisely what survives for the entries.
