# Bloch Sphere Geodesics and the Fubini-Study Metric


---

## Background

Two quantum states that differ only by an overall phase give identical predictions for every measurement, so that phase carries no physical content. The honest space of pure states is therefore not the Hilbert space itself but the set of rays that survive once the phase is divided out, the complex projective space. To do geometry there we need a notion of distance that ignores the phase, and the natural one starts from the overlap of two nearby states and keeps only the part of their difference that moves the ray, discarding the part that merely rotates the phase.

That construction is the Fubini-Study metric. The distance it assigns to two states is the angle $\arccos|\langle\psi|\phi\rangle|$ between the rays they span, so states with large overlap sit close together and orthogonal states sit a quarter turn apart. Reading distance as an angle is more than a figure of speech, since the line element is literally the metric of a curved surface and the shortest paths are arcs along it.

The same line element measures how quickly a state moves under evolution, because a Hamiltonian pushes a state a definite distance per unit time in this geometry. It is the real part of the quantum geometric tensor, which is why it returns in quantum speed limits and in questions of how sensitively a state responds to a parameter. For a single qubit the space of rays is two-dimensional, and the metric works out to that of a round sphere, the familiar Bloch sphere, though it carries a radius that the bare Bloch vector does not make obvious.

This exercise derives the line element directly from the qubit state in the half-angle parametrization, reduces it to the metric of a sphere of radius one half, and measures the geodesic distance from the north pole to a point on the equator, both by integrating along the meridian and from the closed-form overlap formula.

## Mathematical background

The Fubini-Study metric is the distance on the space of rays $\mathbb{CP}^{D-1}$ induced by the Hilbert-space inner product once the global phase is removed. For a normalized $|\psi\rangle$ and a nearby $|\psi + d\psi\rangle$ the line element is

$$
ds^2 = \langle d\psi | d\psi\rangle - |\langle \psi | d\psi\rangle|^2 ,
$$

the squared norm of the variation with its component along $|\psi\rangle$ projected out, which is what makes it phase-invariant. Equivalently the geodesic distance between two rays is the Fubini-Study angle

$$
\gamma(\psi,\phi) = \arccos|\langle\psi|\phi\rangle| ,
$$

so orthogonal states sit a quarter turn apart and the overlap fixes the distance. For a qubit in the half-angle parametrization $|\psi\rangle = \cos(\theta/2)|0\rangle + e^{i\varphi}\sin(\theta/2)|1\rangle$, the line element reduces to $ds^2 = \tfrac14\big(d\theta^2 + \sin^2\theta\, d\varphi^2\big)$, the metric of a sphere of radius $\tfrac12$. The Bloch sphere drawn with unit radius therefore carries twice the Fubini-Study distance, the factor that appears whenever overlaps are turned into angles.

## Exercise

A general single-qubit pure state is

$$
|\psi(\theta, \varphi)\rangle = \cos\frac{\theta}{2}|0\rangle + e^{i\varphi}\sin\frac{\theta}{2}|1\rangle.
$$

**(a)** Starting from the Fubini-Study line element $ds^2_{\mathrm{FS}} = \langle d\psi|d\psi\rangle - |\langle\psi|d\psi\rangle|^2$, show that

$$
ds^2_{\mathrm{FS}} = \frac{1}{4}\bigl(d\theta^2 + \sin^2\!\theta\, d\varphi^2\bigr).
$$

**(b)** Compute the geodesic distance between $|0\rangle$ (north pole) and $|+\rangle = (|0\rangle + |1\rangle)/\sqrt{2}$ (equator).

## Solution

### Part (a): Deriving the metric

**Step 1: Compute $|d\psi\rangle$.**

Differentiating the state:

$$
|d\psi\rangle = -\frac{1}{2}\sin\frac{\theta}{2}\,d\theta\,|0\rangle + e^{i\varphi}\left[\frac{1}{2}\cos\frac{\theta}{2}\,d\theta + i\sin\frac{\theta}{2}\,d\varphi\right]|1\rangle.
$$

**Step 2: Compute $\langle d\psi|d\psi\rangle$ (Term 1).**

The squared norm gives three contributions:

$$
\langle d\psi|d\psi\rangle = \frac{1}{4}\sin^2\frac{\theta}{2}\,d\theta^2 + \frac{1}{4}\cos^2\frac{\theta}{2}\,d\theta^2 + \sin^2\frac{\theta}{2}\,d\varphi^2.
$$

Using $\sin^2 + \cos^2 = 1$:

$$
\langle d\psi|d\psi\rangle = \frac{1}{4}d\theta^2 + \sin^2\frac{\theta}{2}\,d\varphi^2.
$$

**Step 3: Compute $|\langle\psi|d\psi\rangle|^2$ (Term 2).**

The inner product $\langle\psi|d\psi\rangle$ has two contributions:

- From $d\theta$: $\cos\frac{\theta}{2}\cdot(-\frac{1}{2}\sin\frac{\theta}{2}) + \sin\frac{\theta}{2}\cdot\frac{1}{2}\cos\frac{\theta}{2} = 0$. (The real part cancels!)
- From $d\varphi$: $\sin\frac{\theta}{2}\cdot i\sin\frac{\theta}{2} = i\sin^2\frac{\theta}{2}\,d\varphi$.

Therefore:

$$
|\langle\psi|d\psi\rangle|^2 = \sin^4\frac{\theta}{2}\,d\varphi^2.
$$

**Step 4: Combine.**

$$
ds^2_{\mathrm{FS}} = \frac{1}{4}d\theta^2 + \sin^2\frac{\theta}{2}\,d\varphi^2 - \sin^4\frac{\theta}{2}\,d\varphi^2 = \frac{1}{4}d\theta^2 + \sin^2\frac{\theta}{2}\cos^2\frac{\theta}{2}\,d\varphi^2.
$$

Using $\sin^2\frac{\theta}{2}\cos^2\frac{\theta}{2} = \frac{1}{4}\sin^2\theta$:

$$
\boxed{ds^2_{\mathrm{FS}} = \frac{1}{4}\bigl(d\theta^2 + \sin^2\!\theta\,d\varphi^2\bigr).}
$$

This is the metric of a **round sphere of radius** $R = 1/2$. The Bloch sphere has half the radius of the "geographic" unit sphere, the factor $1/2$ arises from the half-angle parametrization of $\mathrm{SU}(2)$.

### Part (b): Geodesic distance $|0\rangle \to |+\rangle$

The north pole is $\theta = 0$ and $|+\rangle$ is at $\theta = \pi/2$, $\varphi = 0$. Along the meridian ($d\varphi = 0$), the geodesic length is:

$$
d_{\mathrm{FS}} = \int_0^{\pi/2} \frac{1}{2}\,d\theta = \frac{\pi}{4}.
$$

Equivalently, using the closed-form Fubini-Study distance:

$$
d_{\mathrm{FS}}(|\psi\rangle, |\phi\rangle) = \arccos|\langle\psi|\phi\rangle|.
$$

$$
|\langle 0|+\rangle| = \frac{1}{\sqrt{2}}, \qquad d_{\mathrm{FS}} = \arccos\frac{1}{\sqrt{2}} = \frac{\pi}{4}. \quad \checkmark
$$

## Symbolic Verification (SymPy)

```python
import sympy as sp

t, p = sp.symbols('theta phi', real=True)

# Term 1: <dÏˆ|dÏˆ>
term1_dtheta = sp.Rational(1, 4)
term1_dphi = sp.sin(t/2)**2

# Term 2: |<Ïˆ|dÏˆ>|Â²
term2_dphi = sp.sin(t/2)**4

# FS metric coefficient for dÏ†Â²
g_phi = sp.trigsimp(term1_dphi - term2_dphi)
print(f"g_Ï†Ï† = {g_phi}")
print(f"g_Ï†Ï† = (1/4)sinÂ²Î¸ : {sp.trigsimp(g_phi - sp.sin(t)**2/4) == 0}")

# Geodesic distance
dist = sp.acos(1/sp.sqrt(2))
print(f"d_FS(|0âŸ©, |+âŸ©) = {dist} = {sp.nsimplify(dist)}")
```

## Numerical Verification

```python
import numpy as np

# Verify d_FS via the overlap formula for several pairs
ket0 = np.array([1, 0], dtype=complex)
ket1 = np.array([0, 1], dtype=complex)
ket_plus = np.array([1, 1], dtype=complex) / np.sqrt(2)

pairs = [
    ("|0âŸ©", "|+âŸ©", ket0, ket_plus, np.pi/4),
    ("|0âŸ©", "|1âŸ©", ket0, ket1, np.pi/2),
    ("|0âŸ©", "|0âŸ©", ket0, ket0, 0.0),
]

for name1, name2, psi, phi, expected in pairs:
    overlap = abs(psi.conj() @ phi)
    d_fs = np.arccos(np.clip(overlap, 0, 1))
    print(f"d_FS({name1}, {name2}) = {d_fs:.4f} (expected {expected:.4f})")
    assert np.isclose(d_fs, expected, atol=1e-10)

print("\nAll assertions passed.")
```

```python
# --- Visualization: the Fubini-Study geodesic on the qubit sphere ---
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(9, 4))

# (a) meridian slice: the geodesic from |0> (north pole) to |+> (equator, theta = pi/2)
ang = np.linspace(0, 2 * np.pi, 200)
ax[0].plot(np.sin(ang), np.cos(ang), color="lightgray", lw=1)
arc = np.linspace(0, np.pi / 2, 60)
ax[0].plot(np.sin(arc), np.cos(arc), color="C0", lw=3)
ax[0].plot(0, 1, "ko", ms=6); ax[0].annotate(r"$|0\rangle$", (0.04, 1.02))
ax[0].plot(1, 0, "ko", ms=6); ax[0].annotate(r"$|+\rangle$", (0.86, 0.07))
ax[0].set_aspect("equal"); ax[0].axis("off")
ax[0].set_title(r"geodesic $|0\rangle\!\to\!|+\rangle$,   $d_{\rm FS}=\pi/4$")

# (b) FS distance arccos|<0|psi(theta)>| = theta/2: linear in the Bloch polar angle
theta = np.linspace(0, np.pi, 100)
d_fs = np.arccos(np.clip(np.abs(np.cos(theta / 2)), 0, 1))
ax[1].plot(theta, d_fs, color="C0", lw=2, label=r"$\arccos|\langle0|\psi(\theta)\rangle|$")
ax[1].plot(theta, theta / 2, "k--", lw=1, label=r"$\theta/2$  (radius-$\frac{1}{2}$ sphere)")
ax[1].plot(np.pi / 2, np.pi / 4, "C3o", ms=7)
ax[1].annotate(r"$|+\rangle$", (np.pi / 2, np.pi / 4),
               textcoords="offset points", xytext=(8, -12))
ax[1].set_xlabel(r"Bloch polar angle $\theta$"); ax[1].set_ylabel(r"$d_{\rm FS}$")
ax[1].set_title("distance is half the polar angle"); ax[1].legend(fontsize=9)

plt.tight_layout()
plt.show()
```

## Takeaway

The Fubini-Study line element is that of a round sphere of radius $1/2$, and the geodesic distance equals $\arccos|\langle\psi|\phi\rangle|$. For the qubit the geometry of states is exactly the geometry of that sphere, with the half-angle parametrization accounting for the radius.
