# Exercise 1.2 — The Data-Processing Inequality for a Dephasing Channel

**Chapter 1: Mathematical Foundations** &nbsp;|&nbsp; *Exercises*

## Motivation

The data-processing inequality is the statement that information is never created by processing.  Sending both members of a pair of states through the same channel can only make them harder to tell apart, $T(\mathcal{E}(\rho), \mathcal{E}(\sigma)) \leq T(\rho,\sigma)$ for every CPTP map $\mathcal{E}$, and this contractivity is what makes the trace distance a meaningful measure of distinguishability at all.

The dephasing channel puts the mechanism in plain sight.  It acts on the Bloch vector as a uniform contraction of the transverse components while leaving the populations alone, so the trace distance can be tracked exactly as coherence is destroyed, and the two states can be watched approaching each other without ever merging.  The same channel is the canonical model of $T_2$ decay, so the calculation is also a statement about how much of a qubit's information survives pure dephasing.

## Exercise Statement

Let $\rho$ and $\sigma$ be the two states of the previous exercise at $\alpha = \pi/2$, namely $\rho = |0\rangle\langle 0|$ and $\sigma = |+\rangle\langle +|$, and let $\mathcal{E}_p$ be the dephasing channel

$$
\mathcal{E}_p(X) = (1-p)\,X + p\,\sigma^z X \sigma^z .
$$

**(a)** Show that $\mathcal{E}_p$ shrinks the $x$ and $y$ Bloch components by $(1-2p)$ and leaves $z$ untouched.

**(b)** Compute $T(\mathcal{E}_p(\rho), \mathcal{E}_p(\sigma))$ and verify the data-processing inequality $T(\mathcal{E}_p(\rho),\mathcal{E}_p(\sigma)) \leq T(\rho,\sigma)$ for all $p \in [0,\tfrac{1}{2}]$.

**(c)** At $p = \tfrac{1}{2}$ the channel is fully dephasing.  What is the trace distance then, and what does that say about distinguishing the two states by any measurement whatsoever?

## Solution

### Part (a): Action on the Bloch vector

The channel has Kraus operators $K_0 = \sqrt{1-p}\,\mathbb{I}$ and $K_1 = \sqrt{p}\,\sigma^z$, which makes it manifestly completely positive.  Trace preservation follows from

$$
K_0^\dagger K_0 + K_1^\dagger K_1 = (1-p)\,\mathbb{I} + p\,(\sigma^z)^2 = (1-p)\,\mathbb{I} + p\,\mathbb{I} = \mathbb{I},
$$

so $\mathcal{E}_p$ is a legitimate channel for every $p \in [0,1]$ and the data-processing inequality must apply to it.

Conjugation by $\sigma^z$ acts on the Pauli basis according to whether the operator commutes or anticommutes with $\sigma^z$.  Since $\sigma^z\sigma^x = -\sigma^x\sigma^z$ and $\sigma^z\sigma^y = -\sigma^y\sigma^z$, while $\sigma^z$ commutes with $\mathbb{I}$ and with itself, and $(\sigma^z)^2 = \mathbb{I}$,

$$
\sigma^z \mathbb{I}\,\sigma^z = \mathbb{I}, \qquad
\sigma^z \sigma^x \sigma^z = -\sigma^x, \qquad
\sigma^z \sigma^y \sigma^z = -\sigma^y, \qquad
\sigma^z \sigma^z \sigma^z = \sigma^z .
$$

Applying $\mathcal{E}_p$ to a general qubit operator $X = \frac{1}{2}(\mathbb{I} + \vec{r}\cdot\vec{\sigma})$,

$$
\mathcal{E}_p(X)
= \tfrac{1}{2}\Bigl[(1-p)\bigl(\mathbb{I} + r_x\sigma^x + r_y\sigma^y + r_z\sigma^z\bigr)
+ p\bigl(\mathbb{I} - r_x\sigma^x - r_y\sigma^y + r_z\sigma^z\bigr)\Bigr]
= \tfrac{1}{2}\Bigl[\mathbb{I} + (1-2p)\,r_x\sigma^x + (1-2p)\,r_y\sigma^y + r_z\sigma^z\Bigr],
$$

where the $\sigma^x$ and $\sigma^y$ terms carry $(1-p) - p = 1-2p$ and the $\sigma^z$ term carries $(1-p) + p = 1$.  Hence

$$
\boxed{(r_x, r_y, r_z) \;\longmapsto\; \bigl((1-2p)\,r_x,\; (1-2p)\,r_y,\; r_z\bigr).}
$$

The Bloch ball is squashed towards the $z$ axis by the factor $1-2p$ while the populations $\langle\sigma^z\rangle$ are left exactly as they were.  At $p = \frac{1}{2}$ the transverse components are erased outright, and at $p = 1$ the channel is the unitary conjugation $X \mapsto \sigma^z X \sigma^z$, a rotation by $\pi$ about $z$.

### Part (b): Trace distance after the channel

At $\alpha = \pi/2$ the two states are $\rho = |0\rangle\langle 0|$ and $\sigma = |+\rangle\langle +|$, with Bloch vectors

$$
\vec{r}_\rho = (0,0,1), \qquad \vec{r}_\sigma = (1,0,0),
$$

so the input trace distance is $T(\rho,\sigma) = \frac{1}{2}|\vec{r}_\rho - \vec{r}_\sigma| = \frac{1}{2}\sqrt{2} = 1/\sqrt{2}$, matching $|\sin(\alpha/2)| = \sin(\pi/4)$ from the previous exercise.

$\vec{r}_\rho$ points along $z$ and part (a) leaves it alone; $\vec{r}_\sigma$ is purely transverse and contracts:

$$
\mathcal{E}_p: \qquad \vec{r}_\rho \longmapsto (0,0,1), \qquad \vec{r}_\sigma \longmapsto (1-2p,\,0,\,0).
$$

Both outputs are still qubit states, so $T = \frac{1}{2}|\Delta\vec{r}|$ continues to apply.  The output separation is $\Delta\vec{r} = (-(1-2p), 0, 1)$, and

$$
\boxed{T\bigl(\mathcal{E}_p(\rho), \mathcal{E}_p(\sigma)\bigr) = \tfrac{1}{2}\sqrt{(1-2p)^2 + 1}.}
$$

On $p \in [0,\frac{1}{2}]$ this falls monotonically from $1/\sqrt{2} \approx 0.7071$ at $p = 0$ to $\frac{1}{2}$ at $p = \frac{1}{2}$, never exceeding the input value $1/\sqrt{2}$.  The data-processing inequality holds, as it must, since post-processing cannot manufacture distinguishability.

The two ingredients of the result pull in opposite directions.  Contracting $\vec{r}_\sigma$ is what shrinks the distance, while the survival of $\vec{r}_\rho$ is what stops the distance from reaching zero, and only the transverse part of the separation is ever attacked.

**Beyond $p = \frac{1}{2}$.**  The boxed formula is valid on all of $[0,1]$ and is symmetric about $p = \frac{1}{2}$, rising back to $1/\sqrt{2}$ at $p = 1$.  There is no contradiction, because at $p = 1$ the channel is a unitary and unitaries preserve the trace distance exactly.  Equality in the data-processing inequality holds at $p = 0$ and $p = 1$ and nowhere else, which is the usual signature: information is lost only where the channel genuinely mixes two Kraus branches.

### Part (c): Full dephasing

At $p = \frac{1}{2}$ the channel is $\mathcal{E}_{1/2}(X) = \frac{1}{2}\bigl(X + \sigma^z X \sigma^z\bigr)$, which deletes every off-diagonal element in the computational basis.  The two states become

$$
\mathcal{E}_{1/2}(\rho) = |0\rangle\langle 0|, \qquad
\mathcal{E}_{1/2}(\sigma) = \frac{\mathbb{I}}{2},
$$

the first unchanged because it was already diagonal, the second reduced to the maximally mixed state because $|+\rangle\langle+|$ carries all of its information in the coherences.  The formula of part (b) gives

$$
\boxed{T = \tfrac{1}{2}\sqrt{0 + 1} = \tfrac{1}{2}.}
$$

The states remain partially distinguishable.  Dephasing preserves the population difference along $z$, and $\rho$ still puts all of its weight on $|0\rangle$ while $\sigma$ has become an even mixture, so measuring $\sigma^z$ still reveals something about which state was prepared.  Through the Helstrom bound, the best single-shot success probability drops from $\frac{1}{2}(1 + 1/\sqrt{2}) \approx 0.854$ before the channel to

$$
p_{\mathrm{correct}} = \tfrac{1}{2}\bigl(1 + \tfrac{1}{2}\bigr) = 0.75
$$

after it, and this now bounds every conceivable measurement on the dephased states, not merely the $\sigma^z$ measurement that suggests itself.  What the channel destroys is coherence rather than distinguishability as such.

## Symbolic Verification (SymPy)

```python
import sympy as sp

p = sp.Symbol('p', real=True)
rx, ry, rz = sp.symbols('r_x r_y r_z', real=True)

I2 = sp.eye(2)
sx = sp.Matrix([[0, 1], [1, 0]])
sy = sp.Matrix([[0, -sp.I], [sp.I, 0]])
sz = sp.Matrix([[1, 0], [0, -1]])

# Trace preservation: K0 = sqrt(1-p) I, K1 = sqrt(p) sz
K0dK0 = (1 - p) * I2
K1dK1 = p * (sz * sz)
print("K0†K0 + K1†K1 =", (K0dK0 + K1dK1).tolist())
assert sp.simplify(K0dK0 + K1dK1 - I2) == sp.zeros(2)
print("E_p is trace preserving for every p ✓")
```

```python
# Part (a): the Bloch map
X = (I2 + rx * sx + ry * sy + rz * sz) / 2
EX = sp.simplify((1 - p) * X + p * sz * X * sz)

for name, pauli in [('r_x', sx), ('r_y', sy), ('r_z', sz)]:
    out = sp.simplify((pauli * EX).trace())
    print(f"{name} -> {out}")

X_pred = (I2 + (1 - 2*p) * rx * sx + (1 - 2*p) * ry * sy + rz * sz) / 2
assert sp.simplify(EX - X_pred) == sp.zeros(2)
print("\n(r_x, r_y, r_z) -> ((1-2p) r_x, (1-2p) r_y, r_z) ✓")
```

```python
# Part (b): trace distance of the outputs, from trace and determinant
rho = sp.Matrix([[1, 0], [0, 0]])                 # |0><0|
sigma = sp.Matrix([[1, 1], [1, 1]]) / 2           # |+><+|

E_rho = sp.simplify((1 - p) * rho + p * sz * rho * sz)
E_sigma = sp.simplify((1 - p) * sigma + p * sz * sigma * sz)
print("E_p(rho)   =", E_rho.tolist())
print("E_p(sigma) =", E_sigma.tolist())

D = sp.simplify(E_rho - E_sigma)
print("trace(D)   =", sp.simplify(D.trace()), "   det(D) =", sp.simplify(D.det()))

# D is traceless and Hermitian, so its eigenvalues are ±sqrt(-det) and T = sqrt(-det)
T_out = sp.simplify(sp.sqrt(-D.det()))
T_in = sp.sqrt(2) / 2
print("\nT(E_p(rho), E_p(sigma)) =", T_out)
assert sp.simplify(T_out - sp.sqrt((1 - 2*p)**2 + 1) / 2) == 0
print("                        =", sp.simplify(sp.sqrt((1 - 2*p)**2 + 1) / 2), " ✓")
print("T at p = 0   :", sp.simplify(T_out.subs(p, 0)), "= input T =", T_in)
print("T at p = 1/2 :", sp.simplify(T_out.subs(p, sp.Rational(1, 2))))
```

```python
# Part (b), continued: monotonicity and the equality set of the DPI
dT = sp.simplify(sp.diff(T_out, p))
print("dT/dp =", dT, "  -> negative on 0 < p < 1/2, so T decreases ✓")
assert sp.simplify(dT.subs(p, sp.Rational(1, 4))) < 0

print("T(E_p(rho), E_p(sigma)) = T(rho, sigma) only at p in",
      sp.solveset(sp.Eq(T_out, T_in), p, domain=sp.S.Reals))

# The DPI itself: T_in^2 - T_out^2 = (1 - (1-2p)^2) / 4 = p(1-p) >= 0 on [0,1]
slack = sp.simplify(T_in**2 - T_out**2)
print("T_in^2 - T_out^2 =", sp.factor(slack), " >= 0 for p in [0,1] ✓")
assert sp.simplify(slack - p * (1 - p)) == 0
```

## Numerical Verification (NumPy)

```python
import numpy as np

I2 = np.eye(2, dtype=complex)
sx = np.array([[0, 1], [1, 0]], dtype=complex)
sy = np.array([[0, -1j], [1j, 0]])
sz = np.array([[1, 0], [0, -1]], dtype=complex)


def dephase(X, p):
    """Apply E_p through its Kraus operators K0 = sqrt(1-p) I, K1 = sqrt(p) sz."""
    K0, K1 = np.sqrt(1 - p) * I2, np.sqrt(p) * sz
    return K0 @ X @ K0.conj().T + K1 @ X @ K1.conj().T


def trace_distance(a, b):
    return 0.5 * np.sum(np.abs(np.linalg.eigvalsh(a - b)))


def bloch(X):
    return np.array([np.real(np.trace(s @ X)) for s in (sx, sy, sz)])


# Part (a) on a random mixed state, not just on the two states of the exercise
rng = np.random.default_rng(7)
r = rng.normal(size=3)
r *= 0.8 / np.linalg.norm(r)
X = 0.5 * (I2 + r[0] * sx + r[1] * sy + r[2] * sz)

print("input Bloch vector :", np.round(r, 6))
for p in [0.0, 0.2, 0.5, 1.0]:
    out = bloch(dephase(X, p))
    pred = np.array([(1 - 2*p) * r[0], (1 - 2*p) * r[1], r[2]])
    print(f"p={p:.1f}: out = {np.round(out, 6)}   predicted = {np.round(pred, 6)}")
    assert np.allclose(out, pred, atol=1e-12)
print("\nBloch contraction by (1-2p) in x,y with z untouched ✓")
```

```python
# Parts (b) and (c): trace distance of the dephased pair
alpha = np.pi / 2
psi = np.array([np.cos(alpha / 2), np.sin(alpha / 2)], dtype=complex)
rho = np.array([[1, 0], [0, 0]], dtype=complex)
sigma = np.outer(psi, psi.conj())

T_in = trace_distance(rho, sigma)
print(f"input:  T(rho, sigma) = {T_in:.6f}   (analytic 1/sqrt(2) = {1/np.sqrt(2):.6f})")
print(f"        p_correct     = {0.5*(1+T_in):.6f}   (analytic {0.5*(1+1/np.sqrt(2)):.6f})\n")

print("   p    r_sigma out          T_out (num)   (1/2)sqrt((1-2p)^2+1)   T_out <= T_in")
for p in [0.0, 0.1, 0.25, 0.4, 0.5]:
    r_out = bloch(dephase(sigma, p))
    T_out = trace_distance(dephase(rho, p), dephase(sigma, p))
    T_an = 0.5 * np.sqrt((1 - 2*p)**2 + 1)
    print(f"{p:5.2f}   {np.array2string(r_out, precision=3, suppress_small=True):18s}"
          f" {T_out:9.6f}      {T_an:9.6f}              {T_out <= T_in + 1e-12}")
    assert np.isclose(T_out, T_an, atol=1e-12)
    assert T_out <= T_in + 1e-12

# Fine grid over the full parameter range of the channel
ps = np.linspace(0.0, 1.0, 501)
T_grid = np.array([trace_distance(dephase(rho, p), dephase(sigma, p)) for p in ps])
assert np.all(T_grid <= T_in + 1e-12)
assert np.allclose(T_grid, 0.5 * np.sqrt((1 - 2*ps)**2 + 1), atol=1e-12)
print(f"\nDPI verified at 501 values of p in [0,1]: max T_out = {T_grid.max():.6f} <= {T_in:.6f} ✓")
print(f"minimum at p = {ps[np.argmin(T_grid)]:.3f}, T = {T_grid.min():.6f}  (analytic 1/2)")
```

```python
# Part (c): the fully dephased pair at p = 1/2
E_rho, E_sigma = dephase(rho, 0.5), dephase(sigma, 0.5)
print("E_{1/2}(rho)   =", np.round(np.real(E_rho), 6).tolist(), " = |0><0|")
print("E_{1/2}(sigma) =", np.round(np.real(E_sigma), 6).tolist(), " = I/2")
assert np.allclose(E_rho, rho, atol=1e-12)
assert np.allclose(E_sigma, I2 / 2, atol=1e-12)

T_half = trace_distance(E_rho, E_sigma)
p_half = 0.5 * (1 + T_half)
print(f"\nT   = {T_half:.6f}   (analytic 1/2)")
print(f"p_correct = {p_half:.6f}   (analytic 0.75), down from {0.5*(1+T_in):.6f} (analytic 0.853553)")
assert np.isclose(T_half, 0.5)
assert np.isclose(p_half, 0.75)

print("\nAll assertions passed. ✓")
```
