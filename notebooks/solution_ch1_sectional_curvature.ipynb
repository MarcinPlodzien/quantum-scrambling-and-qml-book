# Sectional Curvature of SU(2)


---

## Background

Curvature measures how a space bends, and the sectional curvature makes this local by examining one two-dimensional slice at a time. Pick a point and a plane of directions through it, then follow the geodesics that leave in those directions and watch whether they spread apart faster or slower than they would on a flat plane. Positive sectional curvature means nearby geodesics bend back toward each other, the way great circles on a sphere reconverge, while negative curvature means they fly apart.

The unitary group is more than a manifold, since it is also a group, and it carries a metric that looks the same from every point and is unchanged by multiplication from either side. For such a bi-invariant metric the curvature is fixed entirely by the algebra of generators through the commutator. The sectional curvature of the plane spanned by two orthonormal generators $X$ and $Y$ is $K(X,Y) = \tfrac14\|[X,Y]\|^2$, which can never be negative. Generators that fail to commute curve the space, and generators that commute leave it flat.

For $SU(2)$ the three generators are built from the Pauli matrices and stand on equal footing, so every plane returns the same curvature and the group is geometrically a round three-sphere. Working out the commutator of two orthonormal generators pins down the constant and identifies the radius as $\sqrt{2}$. The non-negative curvature carries a physical message, since it means geodesics generated by nearby Hamiltonians can separate at most linearly in time, which limits how quickly single-unitary dynamics alone can drive two trajectories apart.

The notebook evaluates the commutator of the orthonormal Pauli generators and reads off the constant sectional curvature.

## Mathematical background

The radius can be read off from the induced metric in explicit coordinates. Parametrizing $SU(2)$ by Euler angles,

$$
V(\beta,\gamma,\delta) = \begin{pmatrix} e^{i\beta}\cos\gamma & e^{i\delta}\sin\gamma \\ -e^{-i\delta}\sin\gamma & e^{-i\beta}\cos\gamma \end{pmatrix},
$$

the line element $ds^2 = \mathrm{Tr}(dV^\dagger dV)$ is diagonal, with $g_{\gamma\gamma} = 2$, $g_{\beta\beta} = 2\cos^2\gamma$, and $g_{\delta\delta} = 2\sin^2\gamma$. The cross terms vanish because the tangent directions are generated by the orthogonal Pauli matrices, $\mathrm{Tr}(\sigma_i\sigma_j) = 2\delta_{ij}$. Collecting the coefficients,

$$
ds^2 = 2\big(d\gamma^2 + \cos^2\gamma\, d\beta^2 + \sin^2\gamma\, d\delta^2\big),
$$

which is twice the round metric on a unit three-sphere. The overall factor of two carried by the Hilbert-Schmidt metric is what makes the sphere have radius $\sqrt{2}$ rather than one, consistent with the constant curvature $K = 1/R^2 = 1/2$ found from the commutator. The geometry of $SU(2)$ is therefore exactly $S^3$, and evolution under a fixed Hamiltonian moves along its great circles.

## Motivation

Unitary groups form **curved** Riemannian manifolds. The geometry of geodesics on these manifolds governs how physical systems evolve under random or structured Hamiltonian dynamics. A key descriptor of this geometry is the **sectional curvature**, which describes how much geodesics (time-evolution trajectories) tend to deviate from one another.

For Lie groups equipped with a bi-invariant metric (like the Hilbert-Schmidt metric), the sectional curvature is everywhere non-negative. This implies that nearby trajectories generated by slightly different Hamiltonians will slowly drift apart (or oscillate), but they do not exhibit the exponentially fast divergence characteristic of negative curvature. This sets a fundamental continuous-time bound on quantum chaos before many-body effects are considered.

## Preliminaries / Toolbox

**1. The Lie Algebra $\mathfrak{u}(D)$:** The tangent space at the identity consists of skew-Hermitian matrices $X^\dagger = -X$.

**2. The Hilbert-Schmidt Metric:** A natural bi-invariant inner product on the Lie algebra:
$$
g(A, B) = -\mathrm{Tr}(A B).
$$
The minus sign ensures positivity because for skew-Hermitian matrices, $BA$ and $AB$ have negative eigenvalues, so $-\mathrm{Tr}(A^2) = \mathrm{Tr}(A^\dagger A) \geq 0$.

**3. Sectional Curvature:** For a plane spanned by two orthonormal tangent vectors $X, Y \in \mathfrak{u}(D)$ (such that $g(X,X)=1$, $g(Y,Y)=1$, and $g(X,Y)=0$), the sectional curvature of the bi-invariant metric is:
$$
K(X, Y) = \frac{1}{4} \| [X, Y] \|^2_{HS}
$$
where $\| Z \|^2_{HS} = g(Z, Z) = -\mathrm{Tr}(Z^2)$.

**4. Spherical Geometry:** The $n$-dimensional sphere $S^n(R)$ of radius $R$ has constant sectional curvature $K = 1/R^2$ everywhere.

## Exercise

The group $\mathrm{SU}(2)$ is diffeomorphic to the 3-sphere $S^3$.

**(a)** Verify that $X = -i\sigma_x / \sqrt{2}$ and $Y = -i\sigma_y / \sqrt{2}$ are orthonormal in the Hilbert-Schmidt metric.

**(b)** Compute the commutator $[X,Y]$, its norm-squared $\|[X,Y]\|^2_{HS}$, and the resulting sectional curvature $K(X,Y)$.

**(c)** Verify that this matches the expected curvature $K = 1/R^2$ of $S^3$ with radius $R = \sqrt{2}$.

## Solution

### Part (a): Orthonormality

Check the norm of $X$:
$$
\| X \|^2 = -\mathrm{Tr}(X^2) = -\mathrm{Tr}\left(\left(-i\frac{\sigma_x}{\sqrt{2}}\right)^2\right) = -\mathrm{Tr}\left(-\frac{\sigma_x^2}{2}\right) = \frac{\mathrm{Tr}(I_2)}{2} = \frac{2}{2} = 1. \quad \checkmark
$$

Check the norm of $Y$:
$$
\| Y \|^2 = -\mathrm{Tr}(Y^2) = -\mathrm{Tr}\left(-\frac{\sigma_y^2}{2}\right) = \frac{\mathrm{Tr}(I_2)}{2} = 1. \quad \checkmark
$$

Check orthogonality:
$$
g(X, Y) = -\mathrm{Tr}(XY) = -\mathrm{Tr}\left(\left(-i\frac{\sigma_x}{\sqrt{2}}\right)\left(-i\frac{\sigma_y}{\sqrt{2}}\right)\right) = \frac{1}{2}\mathrm{Tr}(\sigma_x \sigma_y) = \frac{1}{2}\mathrm{Tr}(i\sigma_z) = 0. \quad \checkmark
$$

### Part (b): Evaluating the curvature formula

Compute the commutator:
$$
[X, Y] = \left[ -i\frac{\sigma_x}{\sqrt{2}}, -i\frac{\sigma_y}{\sqrt{2}} \right] = -\frac{1}{2} [\sigma_x, \sigma_y] = -\frac{1}{2} (2i\sigma_z) = -i\sigma_z.
$$

Compute its norm-squared:
$$
\| [X,Y] \|^2 = -\mathrm{Tr}\left((-i\sigma_z)^2\right) = -\mathrm{Tr}(-\sigma_z^2) = \mathrm{Tr}(I_2) = 2.
$$

The sectional curvature is therefore:
$$
\boxed{K(X,Y) = \frac{1}{4} \| [X, Y] \|^2 = \frac{1}{4} \times 2 = \frac{1}{2}.}
$$

### Part (c): Geometric interpretation

Because the Lie algebra $\mathfrak{su}(2)$ is 3-dimensional and isotropic under the adjoint action, the sectional curvature is constant for all planes. This implies $\mathrm{SU}(2)$ with the bi-invariant Hilbert-Schmidt metric is isometric to a 3-sphere.

For a sphere of radius $R$, the curvature is $K = 1/R^2$.
$$
K = \frac{1}{2} \implies R^2 = 2 \implies R = \sqrt{2}.
$$

Why $\sqrt{2}$? The standard parametrization $U = x_0 I + i x_1 \sigma_x + i x_2 \sigma_y + i x_3 \sigma_z$ with $\sum x_i^2 = 1$ puts $\mathrm{SU}(2)$ on the unit sphere $S^3(R=1)$ in $\mathbb{R}^4$. The Hilbert-Schmidt metric $g(X,Y) = -\mathrm{Tr}(XY)$ carries an extra factor $\mathrm{Tr}(I_2) = 2$ relative to that flat metric, which rescales every length by $\sqrt{2}$. The sphere therefore has radius $R = \sqrt{2}$ and curvature $K = 1/R^2 = 1/2$, matching part (b).

## Symbolic Verification (SymPy)

```python
import sympy as sp

# Pauli matrices as symbolic 2x2
sx = sp.Matrix([[0, 1], [1, 0]])
sy = sp.Matrix([[0, -sp.I], [sp.I, 0]])
sz = sp.Matrix([[1, 0], [0, -1]])

# Lie algebra elements X = -i*sx/sqrt(2), Y = -i*sy/sqrt(2)
X = -sp.I * sx / sp.sqrt(2)
Y = -sp.I * sy / sp.sqrt(2)

# Check orthonormality: g(A,B) = -Tr(AB)
norm_X2 = -sp.trace(X * X)
norm_Y2 = -sp.trace(Y * Y)
inner_XY = -sp.trace(X * Y)
print(f'||X||^2 = {norm_X2}')
print(f'||Y||^2 = {norm_Y2}')
print(f'g(X,Y) = {inner_XY}')

# Commutator and curvature
comm = X * Y - Y * X
norm_comm2 = -sp.trace(comm * comm)
K = sp.Rational(1, 4) * norm_comm2
print(f'[X,Y] = {comm}')
print(f'||[X,Y]||^2 = {norm_comm2}')
print(f'K(X,Y) = {K}')
assert K == sp.Rational(1, 2)
print(f'Verified: K = 1/2 = 1/R^2 with R = sqrt(2)')
```

## Numerical Verification

## Takeaway

The bi-invariant curvature formula $K(X,Y)=\tfrac14\|[X,Y]\|^2$ gives $K=1/2$ for every plane in $\mathfrak{su}(2)$, identifying $SU(2)$ with a 3-sphere of radius $\sqrt{2}$. Non-negative curvature means geodesics generated by nearby Hamiltonians separate at most linearly rather than exponentially.

```python
import numpy as np
from scipy.linalg import expm

sx = np.array([[0, 1], [1, 0]], dtype=complex)
sy = np.array([[0, -1j], [1j, 0]], dtype=complex)
sz = np.array([[1, 0], [0, -1]], dtype=complex)

X = -1j * sx / np.sqrt(2)
Y = -1j * sy / np.sqrt(2)

def hs_inner(A, B):
    return -np.trace(A @ B).real

norm_X2 = hs_inner(X, X)
norm_Y2 = hs_inner(Y, Y)
inner_XY = hs_inner(X, Y)

print("Lie algebra orthonormality:")
print(f"  ‖X‖²   = {norm_X2:.2f}")
print(f"  ‖Y‖²   = {norm_Y2:.2f}")
print(f"  g(X,Y) = {inner_XY:.2f}")

comm = X @ Y - Y @ X
norm_comm2 = hs_inner(comm, comm)
K_val = 0.25 * norm_comm2

print(f"\nSectional curvature:")
print(f"  [X,Y]    = \n{np.round(comm, 2)}")
print(f"  ‖[X,Y]‖² = {norm_comm2:.2f}")
print(f"  K(X,Y)   = {K_val:.2f}")

assert np.isclose(K_val, 0.5), "Curvature should be 1/2"
print("\nConstant sectional curvature K = 1/2 of SU(2) confirmed.")
```

```python
# --- Visualization: positive curvature => nearby Hamiltonians separate at most linearly ---
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

H = np.array([[1, 0], [0, -1]], dtype=complex)     # base Hamiltonian (sigma_z)
V = np.array([[0, 1], [1, 0]], dtype=complex)      # perturbation (sigma_x)
eps = 0.15
ts = np.linspace(0, 20, 400)
sep = np.array([np.linalg.norm(expm(-1j * H * t) - expm(-1j * (H + eps * V) * t)) for t in ts])

fig, ax = plt.subplots(figsize=(6.6, 4))
ax.plot(ts, sep, color="C0", lw=2, label=r"$\|U_1(t)-U_2(t)\|$ on $SU(2)$")
ax.plot(ts, eps * ts, "k--", lw=1, label=r"linear envelope $\varepsilon\,t$")
ax.plot(ts, 0.2 * np.exp(0.15 * ts), ls=":", color="C3", lw=1.4,
        label="what negative curvature would give")
ax.set_xlabel(r"time $t$"); ax.set_ylabel("trajectory separation")
ax.set_title(r"$K=\frac{1}{2}>0$: bounded separation (no exponential blow-up)")
ax.set_ylim(0, 3); ax.set_xlim(0, 20); ax.legend(fontsize=9, loc="upper left")

plt.tight_layout()
plt.show()
```
