# Dimension of the Lie Algebras $\mathfrak{u}(D)$ and $\mathfrak{su}(D)$


---

## Background

Every continuous evolution of a $D$-dimensional quantum system is generated by a Hermitian Hamiltonian, and pulling out the factor of $i$ turns these generators into skew-Hermitian matrices, the elements of the Lie algebra $\mathfrak{u}(D)$. Keeping only the traceless ones gives $\mathfrak{su}(D)$, the algebra attached to the special unitary group, whose elements generate the unitaries of unit determinant. Asking how large this algebra is amounts to asking how many independent generators it holds, which comes down to counting the real parameters left once the defining constraints are imposed.

The counting is direct. Skew-Hermiticity ties each below-diagonal entry to its mirror image above and forces every diagonal entry to be purely imaginary, so the free real parameters number $D^2$. Requiring the trace to vanish removes exactly one more, since that is a single condition on the imaginary diagonal, leaving $D^2 - 1$ generators for $\mathfrak{su}(D)$. For a single qubit this recovers the three Pauli generators, as it should.

What matters is how the count grows with system size. For $N$ qubits the dimension is $D = 2^N$, and the generator count becomes $4^N - 1$, an exponential in the number of qubits. The space of directions a generic circuit can move in is therefore enormous, and that very abundance is what dilutes the signal in variational optimization. When a circuit randomizes enough to imitate the first two moments of the unitary group, the variance of a cost-function gradient shrinks like $4^{-N}$, and the training landscape flattens into a barren plateau where the gradients are too small to follow.

This exercise carries out the parameter count for $\mathfrak{u}(D)$ and $\mathfrak{su}(D)$, confirms the $4^N - 1$ scaling for qubits, and connects it to the barren plateau.

## Mathematical background

The dimensions follow from counting independent generators. The algebra $\mathfrak{u}(D)$ consists of skew-Hermitian matrices $X^\dagger = -X$. A complex $D\times D$ matrix carries $2D^2$ real parameters, and the skew-Hermitian condition is a real-linear constraint that halves them,

$$
\dim_{\mathbb{R}}\mathfrak{u}(D) = D^2 .
$$

The diagonal entries are pure imaginary ($D$ real parameters) and the $\binom{D}{2}$ off-diagonal pairs each carry two ($D^2 - D$ parameters), summing to $D^2$. Imposing tracelessness $\mathrm{Tr}\,X = 0$ removes one direction,

$$
\dim_{\mathbb{R}}\mathfrak{su}(D) = D^2 - 1 .
$$

For a single qubit this gives $\dim\mathfrak{su}(2) = 3$, the three Pauli generators. These counts set the size of the space a universal gate set must fill, and they reappear directly in the dimension factors of Weingarten averages and barren-plateau gradient variances.

## Exercise

**(a)** Count the real degrees of freedom in a $D \times D$ skew-Hermitian matrix and show $\dim_\mathbb{R}\, \mathfrak{u}(D) = D^2$.

**(b)** The tracelessness constraint $\mathrm{Tr}(X) = 0$ defines $\mathfrak{su}(D) \subset \mathfrak{u}(D)$. Show $\dim_\mathbb{R}\, \mathfrak{su}(D) = D^2 - 1$.

**(c)** For $N$ qubits ($D = 2^N$), how does $\dim\,\mathfrak{su}(D)$ scale with $N$? Relate to the barren plateau phenomenon.

## Solution

### Part (a): Dimension of $\mathfrak{u}(D)$

The Lie algebra $\mathfrak{u}(D)$ consists of all skew-Hermitian matrices: $X = -X^\dagger$. This constraint relates entries pairwise.

**Diagonal entries** ($X_{jj}$): The constraint $X_{jj} = -\overline{X_{jj}}$ forces each diagonal entry to be purely imaginary: $X_{jj} = i\theta_j$ with $\theta_j \in \mathbb{R}$. This gives **$D$ real parameters**.

**Off-diagonal entries** ($j < k$): The upper triangle has $D(D-1)/2$ complex entries, each with 2 real parameters. The lower triangle is determined by $X_{kj} = -\overline{X_{jk}}$. This gives **$D(D-1)$ real parameters**.

**Total:**

$$
\boxed{\dim_\mathbb{R}\,\mathfrak{u}(D) = D + D(D-1) = D^2.}
$$

### Part (b): Dimension of $\mathfrak{su}(D)$

The special unitary algebra $\mathfrak{su}(D)$ imposes the additional constraint $\mathrm{Tr}(X) = 0$. Since

$$
\mathrm{Tr}(X) = \sum_{j=1}^D X_{jj} = i\sum_{j=1}^D \theta_j,
$$

the trace vanishes if and only if $\sum_j \theta_j = 0$. This is a **single real constraint** on the $D$ imaginary diagonal parameters. The off-diagonal entries are unaffected.

$$
\boxed{\dim_\mathbb{R}\,\mathfrak{su}(D) = D^2 - 1.}
$$

This matches the familiar count: $\mathrm{SU}(2)$ has $4 - 1 = 3$ generators (the Pauli matrices).

### Part (c): Exponential scaling and barren plateaus

For $N$ qubits, $D = 2^N$ and

$$
\dim\,\mathfrak{su}(2^N) = 4^N - 1.
$$

This exponential growth has a direct physical consequence: the **gradient variance** in a variational quantum circuit that forms a 2-design scales as

$$
\mathrm{Var}\!\left[\frac{\partial C}{\partial\theta}\right] = O\!\left(\frac{1}{4^N}\right).
$$

The gradient is exponentially small because the cost function "lives" on a manifold with $4^N - 1$ dimensions, the gradient signal is spread over an exponentially large space. This is the **barren plateau** phenomenon.

## Symbolic Verification (SymPy)

```python
import sympy as sp

D = sp.Symbol('D', positive=True, integer=True)

dim_uD = D**2
dim_suD = D**2 - 1

# Verify by components
diag_params = D
off_diag_params = D*(D-1)
total = sp.expand(diag_params + off_diag_params)
print(f"Diagonal: {diag_params}, Off-diagonal: {off_diag_params}")
print(f"Total: {total} = D² = {sp.expand(D**2)}")
assert total == D**2
print(f"dim u(D) = {dim_uD}, dim su(D) = {dim_suD}")
```

---
## Numerical Verification

```python
import numpy as np

for D_val in [2, 4, 8, 16]:
    # Generate a random skew-Hermitian matrix
    A = np.random.randn(D_val, D_val) + 1j*np.random.randn(D_val, D_val)
    X = (A - A.conj().T) / 2  # Project to skew-Hermitian

    # Verify skew-Hermiticity
    assert np.allclose(X, -X.conj().T)

    # Count real parameters
    dim_u = D_val**2
    dim_su = D_val**2 - 1
    N_qubits = int(np.log2(D_val))

    print(f"D={D_val:3d} ({N_qubits}q): dim u(D) = {dim_u:6d}, "
          f"dim su(D) = {dim_su:6d} = 4^{N_qubits}−1 = {4**N_qubits - 1}")
    assert dim_su == 4**N_qubits - 1

print("\nExponential growth: dim su(2^N) = 4^N − 1.")
```

---
## Takeaway

Skew-Hermiticity gives $\dim_\mathbb{R}\,\mathfrak{u}(D) = D^2$, and the single tracelessness constraint gives $\dim_\mathbb{R}\,\mathfrak{su}(D) = D^2 - 1$. For $N$ qubits this is $4^N - 1$, the exponentially large manifold over which variational gradients are spread in a barren plateau.
