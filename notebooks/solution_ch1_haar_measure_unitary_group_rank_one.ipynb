# Haar Invariance on $U(1)$


---

## Background

To average a quantity over random unitaries we first need a sensible notion of what random means on the group, and that is supplied by a measure that weighs subsets of the group. The natural choice is the one that does not change when the group acts on itself by multiplication, called the Haar measure. Invariance under this action is exactly what makes an average meaningful, since it guarantees that relabeling by any fixed unitary leaves the average untouched. On a compact group this invariant measure exists and is unique once the total weight is fixed to one.

The simplest instance is the circle group $U(1)$, whose elements are phases $e^{i\theta}$. Here the Haar measure is the uniform density $d\theta/(2\pi)$, and one can see why invariance forces this form. A density on the circle expands in Fourier modes, and rotating the circle multiplies every nonzero mode by a phase. For the density to come back to itself under all rotations, those modes have to vanish, which leaves only the constant term that normalization then sets.

The same argument reaches into higher dimensions through the determinant. If $U$ is drawn from the Haar measure on $U(D)$, multiplying it by a global phase rotates $\det U$ around the unit circle, and Haar invariance says this rotation cannot change the distribution. The phase of the determinant is therefore forced to be uniform, a simple signature of Haar randomness that persists in any dimension.

The notebook samples random unitaries and compares the distribution of the determinant phase against the uniform law.

## Mathematical background

On the full group $U(D)$ the defining property is invariance under multiplication from either side,

$$
d\mu(U) = d\mu(VU) = d\mu(UV), \qquad \forall\, V \in U(D).
$$

Because $U(D)$ is compact, this measure exists, is unique, and can be normalized to $\int_{U(D)} d\mu(U) = 1$. Left invariance is what guarantees that a Haar-random unitary applied to any fixed reference state produces a state $|\psi\rangle = U|\psi_0\rangle$ whose distribution is the uniform measure on the sphere $S^{2D-1}$, independent of $|\psi_0\rangle$. A concrete way to sample is to draw a complex Ginibre matrix $G$ with independent Gaussian entries and take the $Q$ factor of its QR decomposition, with the diagonal phases of $R$ fixed so that the result is uniform. The $U(1)$ case worked out in this notebook is the smallest instance of the same invariance argument, with the uniform phase density $d\theta/(2\pi)$ as its unique solution.

## Motivation

The **Haar measure** is the unique translation-invariant probability measure on a compact Lie group. Its existence and uniqueness underpin every Haar-random construction in this book, from random matrix theory to unitary designs. The circle group $U(1)$ is the simplest compact Lie group and provides the clearest illustration of these concepts: uniqueness follows from a Fourier analysis argument, and the connection to determinant phases generalizes directly to $U(D)$.

## Exercise

The circle group $U(1) = \{e^{i\theta} : \theta \in [0, 2\pi)\}$ has Haar measure $d\mu = d\theta/(2\pi)$.

**(a)** Verify translation invariance: $\int f(e^{i(\theta+\alpha)})\,\frac{d\theta}{2\pi} = \int f(e^{i\theta})\,\frac{d\theta}{2\pi}$.

**(b)** Prove uniqueness: any continuous translation-invariant density $p(\theta)$ must equal $1/(2\pi)$.

**(c)** For Haar-random $U \in U(D)$, show that the phase of $\det(U)$ is uniformly distributed on $[0, 2\pi)$.

## Solution

### Part (a): Translation invariance

Under the substitution $\theta' = \theta + \alpha$, $d\theta' = d\theta$, and the domain $[0, 2\pi)$ maps to $[\alpha, 2\pi + \alpha)$. Since the integrand $f(e^{i\theta'})$ is $2\pi$-periodic in $\theta'$, the integral over any interval of length $2\pi$ is the same:

$$
\int_0^{2\pi} f(e^{i(\theta+\alpha)})\frac{d\theta}{2\pi} = \int_\alpha^{2\pi+\alpha} f(e^{i\theta'})\frac{d\theta'}{2\pi} = \int_0^{2\pi} f(e^{i\theta'})\frac{d\theta'}{2\pi}. \quad \checkmark
$$

This is the defining property of the Haar measure: it is invariant under group multiplication (here: phase shifts).

### Part (b): Uniqueness via Fourier analysis

Suppose $p(\theta)$ is a translation-invariant density. Expand in Fourier modes:

$$
p(\theta) = \sum_{n=-\infty}^{\infty} c_n\, e^{in\theta}.
$$

Translation invariance requires $\int f(\theta + \alpha)\, p(\theta)\, d\theta = \int f(\theta)\, p(\theta)\, d\theta$ for all $\alpha$ and all test functions $f$. This implies:

$$
e^{in\alpha}\, c_n = c_n \qquad \forall\, \alpha.
$$

For $n \neq 0$, we can choose $\alpha$ such that $e^{in\alpha} \neq 1$, forcing $c_n = 0$. Only $c_0$ survives.

Normalization $\int_0^{2\pi} p(\theta)\,d\theta = 1$ fixes $c_0 = 1/(2\pi)$. Therefore:

$$
\boxed{p(\theta) = \frac{1}{2\pi}.}
$$

The Haar measure is the **unique** translation-invariant measure on $U(1)$.

### Part (c): Determinant phase is uniform

For $U \in U(D)$, write $\det(U) = e^{i\Phi}$ where $\Phi = \sum_k \varphi_k$ is the sum of eigenphases.

Left multiplication $U \to e^{i\alpha}U$ shifts every eigenphase by $\alpha$, so $\det(e^{i\alpha}U) = e^{iD\alpha}\det(U)$. This shifts $\Phi \to \Phi + D\alpha$.

Since the Haar measure is left-invariant, the distribution of $\Phi + D\alpha$ equals the distribution of $\Phi$ for all $\alpha$. By the uniqueness result of part (b) (applied to the group $U(1)$ acting on the determinant phase), $\Phi$ must be uniformly distributed on $[0, 2\pi)$.

## Symbolic Verification (SymPy)

```python
import sympy as sp

t, a, n = sp.symbols('theta alpha n', real=True)

# Fourier coefficient constraint: e^{inα} c_n = c_n
constraint = sp.exp(sp.I*n*a) - 1
print(f"For n≠0, choose α=2π/(3n): e^{{inα}} − 1 = {sp.simplify(constraint.subs(a, 2*sp.pi/(3*n)))}")

# Translation invariance of ∫ e^{imθ} dθ/2π
for m in range(-3, 4):
    orig = sp.integrate(sp.exp(sp.I*m*t), (t, 0, 2*sp.pi)) / (2*sp.pi)
    shifted = sp.integrate(sp.exp(sp.I*m*(t+a)), (t, 0, 2*sp.pi)) / (2*sp.pi)
    print(f"  m={m:+d}: ∫ e^{{imθ}} = {sp.simplify(orig)}, ∫ e^{{im(θ+α)}} = {sp.simplify(shifted)}")
```

## Numerical Verification

## Figure: Sampled $U(1)$ phase is uniform

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import unitary_group

rng = np.random.default_rng(0)
D = 4
n_samples = 20000
phases = np.array([
    np.angle(np.linalg.det(unitary_group.rvs(D, random_state=rng))) % (2 * np.pi)
    for _ in range(n_samples)
])

fig, ax = plt.subplots(figsize=(6.4, 4.0))
ax.hist(phases, bins=40, range=(0, 2 * np.pi), density=True,
        color='steelblue', edgecolor='white', alpha=0.85,
        label='sampled phases')
ax.axhline(1 / (2 * np.pi), color='crimson', lw=2,
           label=r'Haar density $1/2\pi$')
ax.set_xlabel(r'phase of $\det U \in [0, 2\pi)$')
ax.set_ylabel('probability density')
ax.set_title(r'Determinant phase of Haar-random $U(4)$ is uniform')
ax.set_xticks([0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi])
ax.set_xticklabels(['0', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])
ax.set_ylim(0, 0.25)
ax.legend()
plt.tight_layout()
plt.show()
```

## Takeaway

Invariance alone pins down the Haar measure on $U(1)$ as the uniform density $1/(2\pi)$. The same argument, applied to the determinant, shows that $\det U$ has uniform phase for Haar-random $U \in U(D)$.

```python
import numpy as np
from scipy.stats import unitary_group, kstest

np.random.seed(42)

# Verify det(U) phase is uniform for several D
for D in [2, 4, 8]:
    phases = []
    for _ in range(5000):
        U = unitary_group.rvs(D)
        det_phase = np.angle(np.linalg.det(U)) % (2*np.pi)
        phases.append(det_phase / (2*np.pi))  # normalize to [0,1)

    # KS test against uniform[0,1]
    stat, pval = kstest(phases, 'uniform')
    print(f"D={D}: KS statistic = {stat:.4f}, p-value = {pval:.4f} "
          f"[{'uniform' if pval > 0.01 else 'REJECT'}]")
    assert pval > 0.01

print("\nDeterminant phase is uniform for all D.")
```
