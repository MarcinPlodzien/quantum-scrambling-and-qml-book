# SFF Plateau Value for the GUE


---

## Background

The spectral form factor is built from the trace $\mathrm{Tr}\,e^{-iHt}$, the sum of phases $e^{-iE_n t}$ running over the energy levels of a Hamiltonian $H$. Its magnitude squared, $K(t) = |\mathrm{Tr}\,e^{-iHt}|^2$, expands into a double sum over pairs of levels weighted by their energy differences, which is why it serves as a probe of how the levels are correlated. Quantum chaos leaves a characteristic imprint on these correlations, and the form factor is one of the simplest places to read it.

At $t=0$ every phase aligns and the $D^2$ terms add coherently, giving $K(0) = D^2$ for a spectrum of $D$ levels. As time advances the off-diagonal terms, carrying nonzero energy differences, oscillate and interfere away. Only the $D$ diagonal terms, where a level is paired with itself, contribute a steady value, so the long-time average settles at $\overline{K} = D$. The ratio $K(0)/\overline{K} = D$ records how much coherence was present at the start.

Between these two ends the averaged form factor traces a recognizable shape. It first dips below its initial value, then climbs along a linear ramp before flattening into the plateau at $D$. The ramp is the fingerprint of level repulsion, the tendency of chaotic spectra to avoid near-degeneracies, while the plateau height counts levels and nothing more.

The notebook derives the plateau value by time-averaging the double sum and isolating the diagonal terms, then samples random Hamiltonians from the Gaussian unitary ensemble to show how the dip and ramp give way to the plateau pinned at $K = D$.

## Mathematical background

It helps to write the form factor as the squared modulus of a complex-time partition function. With $Z(\beta + it) = \mathrm{Tr}\,e^{-(\beta+it)H} = \sum_n e^{-(\beta+it)E_n}$, the spectral form factor is

$$
K(\beta, t) = |Z(\beta+it)|^2 = \sum_{m,n} e^{-\beta(E_m+E_n)}\,e^{-it(E_m-E_n)} .
$$

Setting $\beta = 0$ isolates the spectral content. The double sum splits into a diagonal part ($m=n$), which contributes the constant $D$, and an off-diagonal part that depends on the distribution of level differences $E_m - E_n$. For an uncorrelated (Poisson) spectrum the off-diagonal phases average to a fast decay with no ramp. For a chaotic spectrum the two-level correlations of the unitary ($\beta = 2$) ensemble instead produce a linear-in-time ramp, $K(t) \propto t$, that rises until the Heisenberg time $t \sim D$ and then saturates at the plateau $\overline{K} = D$. The ramp is the time-domain image of level repulsion, and the plateau height counts the number of levels, so the whole dip-ramp-plateau curve is fixed by the spectral statistics alone and is blind to the choice of operator.

## Motivation

The **spectral form factor** (SFF) $K(t) = |Z(it)|^2 = |\mathrm{Tr}(e^{-iHt})|^2$ is the spectral diagnostic of quantum chaos. It probes pair correlations of energy levels. The SFF exhibits three regimes: a **slope** (dip below $D^2$), a **ramp** (linear in $t$, reflecting level repulsion), and a **plateau** at $K = D$. This exercise derives the plateau value from time-averaging and relates it to the dephasing of off-diagonal terms.

## Preliminaries / Toolbox

Before diving into the solution, recall the following concepts:

**1. Spectral Form Factor (SFF):** Provides a proxy for the analytic continuation of the partition function to study late-time chaos geometry. $K(t) = |Z(i t)|^2 = |\mathrm{Tr}(e^{-i H t})|^2 = \sum_{m,n} e^{-i(E_m - E_n)t}$.

**2. Level Repulsion:** In chaotic models (like GUE), eigenvalues repel each other ($P(\Delta E) \sim \Delta E^\beta \to 0$). They never perfectly cross.

**3. Time Averaging:** Off-diagonal terms $e^{-i(E_m - E_n)t}$ strongly oscillate and vanish under time average, isolating diagonal contributions.

**4. Coherence vs Dephasing:** $K(0) = D^2$ shows full coherence, while $K \to D$ indicates full dephasing into mutually orthogonal eigenvectors.

## Exercise

For a $D \times D$ GUE Hamiltonian with eigenvalues $\{E_n\}$: $K(t) = \sum_{m,n} e^{-it(E_m - E_n)}$.

**(a)** Show the long-time average $\overline{K} = D$.

**(b)** Explain why $K_{\mathrm{plateau}} = D$ (not $D^2$) and interpret the ratio $K(0)/\overline{K}$.

## Solution

### Part (a): Time-averaging the SFF

Time-average each term separately:

$$
\overline{K} = \lim_{T\to\infty}\frac{1}{T}\int_0^T K(t)\,dt = \sum_{m,n} \lim_{T\to\infty} \frac{1}{T}\int_0^T e^{-i(E_m - E_n)t}\,dt.
$$

**Diagonal terms** ($m = n$): The integrand is 1. Each contributes 1. There are $D$ such terms.

**Off-diagonal terms** ($m \neq n$): Let $\omega_{mn} = E_m - E_n \neq 0$. Then:

$$
\frac{1}{T}\int_0^T e^{-i\omega_{mn} t}\,dt = \frac{1 - e^{-i\omega_{mn} T}}{i\omega_{mn} T} \to 0 \quad \text{as } T \to \infty.
$$

GUE level repulsion ensures $\omega_{mn} \neq 0$ with probability 1 (eigenvalues don't coincide).

$$
\boxed{\overline{K} = D.}
$$

### Part (b): Interpretation

At $t = 0$: all $D^2$ terms contribute coherently, $K(0) = D^2$.

At late times: the $D(D-1)$ off-diagonal terms dephase (oscillate with incommensurate frequencies and cancel upon averaging), leaving only the $D$ diagonal terms.

The ratio:

$$
\frac{K(0)}{\overline{K}} = \frac{D^2}{D} = D
$$

quantifies the **coherent enhancement** at $t = 0$. This dephasing from $D^2$ to $D$ is the spectral signature of quantum chaos, the loss of coherence between energy eigenstates over time.

## Symbolic Verification (SymPy)

```python
import sympy as sp

D, t = sp.symbols('D t', positive=True)

# SFF: K(t) = |Tr(e^{-iHt})|^2 = sum_{m,n} e^{-i(E_m - E_n)t}
# Time average: <K> = sum_{m,n} delta(E_m, E_n) * 1
# For non-degenerate spectrum: <K> = D (only diagonal terms survive)
print('Spectral Form Factor time average:')
print(f'  K(0) = |Tr(I)|^2 = D^2')
print(f'  <K>_T = D  (diagonal terms only, non-degenerate)')
print(f'  Ratio K(0)/<K> = D')

# Numerical check
for n in range(2, 8):
    d = 2**n
    print(f'  N={n}: K(0)={d**2}, <K>={d}, ratio={d}')
```

---
## Numerical Verification

```python
import numpy as np

np.random.seed(42)

for D in [4, 8, 16, 32]:
    A = np.random.randn(D,D) + 1j*np.random.randn(D,D)
    H = (A + A.conj().T) / (2*np.sqrt(D))
    E = np.linalg.eigvalsh(H)

    # SFF at 1000 time points and average
    times = np.linspace(10, 500, 1000)
    K_vals = [abs(np.sum(np.exp(-1j*E*t)))**2 for t in times]
    K_bar = np.mean(K_vals)

    print(f"D={D:3d}: K(0)={D**2:5d},  K̄={K_bar:8.2f},  K̄/D={K_bar/D:.3f}")
    assert abs(K_bar/D - 1) < 0.15

print("\nSFF plateau at K̄ = D confirmed.")
```

---
## Visual Pedagogical Proof

```python
import matplotlib.pyplot as plt

# A single GUE realization is too noisy to expose the ramp; averaging the SFF
# over many independent samples reveals the dip-ramp-plateau structure, with the
# linear ramp set by level repulsion and the plateau pinned at K = D.
np.random.seed(45)
D = 256
reps = 200
times_log = np.logspace(-1, 3, 250)

K_sum = np.zeros(len(times_log))
for _ in range(reps):
    A = np.random.randn(D, D) + 1j*np.random.randn(D, D)
    H = (A + A.conj().T) / (2*np.sqrt(D))
    E = np.linalg.eigvalsh(H)
    K_sum += np.abs(np.exp(-1j*np.outer(times_log, E)).sum(axis=1))**2
K = K_sum / reps

plt.rcParams['figure.dpi'] = 150
plt.figure(figsize=(8, 4.5))
plt.loglog(times_log, K, color='royalblue', linewidth=1.8,
           label=f'SFF $K(t)$ (GUE average, {reps} samples)')
plt.axhline(D, color='crimson', linestyle='--', linewidth=2, label=r'Plateau $\bar{K} = D$')
plt.axhline(D**2, color='black', linestyle=':', label=r'Initial $K(0) = D^2$')

plt.xlabel(r"Time $t$")
plt.ylabel(r"Spectral Form Factor $K(t)$")
plt.title(f"SFF for GUE ($D={D}$): Slope, Ramp, and Plateau")
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

## Takeaway

The long-time average of the spectral form factor equals $D$, the number of diagonal terms that outlive dephasing, and the ratio $K(0)/\overline{K} = D$ measures the coherence present at $t = 0$. The plot shows the slope and ramp giving way to a plateau pinned at $K = D$.

## Connections

The ramp of the spectral form factor is the time-domain image of the level rigidity that the [GOE spacing distribution](../ch2/solution_ch2_goe_spacing.ipynb) measures statically, so the same two-level correlations appear once in a histogram of gaps and once in a curve against time.
